```python
df1 = pd.read_csv("Data/Stocks/" + "삼성전자.csv")
df1["Date"] = pd.to_datetime(df1["Date"])

df2 = pd.read_csv("Data/Stocks/" + "현대차.csv")
df2["Date"] = pd.to_datetime(df2["Date"])

df3 = pd.read_csv("Data/Stocks/" + "카카오.csv")
df3["Date"] = pd.to_datetime(df3["Date"])

df4 = pd.read_csv("Data/Stocks/" + "한화오션.csv")
df4["Date"] = pd.to_datetime(df4["Date"])
```

```python
df1 = df1[["Date", "Close"]]
df2 = df2[["Date", "Close"]]
df3 = df3[["Date", "Close"]]
df4 = df4[["Date", "Close"]]

a = df1.merge(df2, on = "Date")
b = a.merge(df3, on = "Date")
c = b.merge(df4, on = "Date", suffixes = ("_left", "_right"))
c.set_index("Date")
c.columns = ["Date", "삼성전자", "현대차", "카카오", "한화오션"]
c = c.set_index("Date")
# del c["Date"]
c
```

```python
daily_ret = c.pct_change()
annual_ret = daily_ret.mean() * 252
daily_cov = daily_ret.cov()
annual_cov = daily_cov * 252
```

```python
# 시뮬레이션
import pandas as pd
import numpy as np

total = []

for n in range(10000):
    weights = np.random.random(4)
    weights = weights / np.sum(weights)

    returns = np.dot(weights, annual_ret)
    risk = np.sqrt(np.dot(weights.T, np.dot(annual_cov, weights)))
    sharpe_ratio = (returns - 0.025) / risk 
    
    total.append([returns, risk, sharpe_ratio] + list(weights))

df = pd.DataFrame(total, columns = ['Returns', 'Risk', "Sharpe", "삼성전자", "현대차", "카카오", "한화오션"])
df

# 각 비중대로 샀을때 수익률은 이렇고 위험률은 이렇다
```

# 문제 5

```python
from scipy.stats import norm

def calculate_kelly_fraction(p: float, b: float) -> float:
    """
    켈리 방정식을 사용하여 최적의 베팅 비율을 계산합니다.

    Args:
        p (float): 승리 확률 (0과 1 사이)
        b (float): 배당률 (손익비, 예: 1000원 걸어 2000원 순수익 시 b=2)

    Returns:
        float: 최적의 베팅 비율 f*. 0보다 작으면 0을 반환.
    """
    if not (0 <= p <= 1):
        raise ValueError("승리 확률(p)은 0과 1 사이의 값이어야 합니다.")
    if b <= 0:
        raise ValueError("배당률(b)은 0보다 커야 합니다.")
        
    q = 1 - p
    f_star = (b * p - q) / b
    
    return max(0, f_star)

# 위에서 들었던 예시로 계산해보기
p_coin = 0.3
b_coin = 3
kelly_fraction = calculate_kelly_fraction(p_coin, b_coin)

print(f"승리 확률: {p_coin}, 배당률: {b_coin}")
print(f"최적의 베팅 비율(켈리 비율): {kelly_fraction:.2%}") 

def calculate_asymmetric_kelly(p: float, win_rate: float, loss_rate: float) -> float:
    """
    비대칭 손익률을 위한 켈리 방정식을 계산합니다.

    Args:
        p (float): 승리 확률 (0과 1 사이)
        win_rate (float): 승리 시 베팅액 대비 수익률 (예: 20% 수익 시 0.2)
        loss_rate (float): 패배 시 베팅액 대비 손실률 (예: 10% 손실 시 0.1)

    Returns:
        float: 최적의 베팅 비율 f*. 0보다 작으면 0을 반환.
    """
    if not (0 <= p <= 1):
        raise ValueError("승리 확률(p)은 0과 1 사이의 값이어야 합니다.")
    if win_rate <= 0 or loss_rate <= 0:
        raise ValueError("수익률(win_rate)과 손실률(loss_rate)은 0보다 커야 합니다.")

    q = 1 - p

    # 기대 수익률이 0보다 작거나 같으면 베팅하지 않음
    # (p * win_rate) > (q * loss_rate) 이어야 베팅의 가치가 있음
    if (p * win_rate) <= (q * loss_rate):
        return 0.0

    # 비대칭 손익 켈리 공식
    # f* = (p*w - q*l) / (w*l)
    numerator = p * win_rate - q * loss_rate
    denominator = win_rate * loss_rate
    
    f_star = numerator / denominator
    
    return max(0, f_star)

# 예시: 승률 60%, 이기면 20% 얻고, 지면 10% 잃는 경우
p_custom = 0.6
w_custom = 0.2  # 20% 수익
l_custom = 0.1  # 10% 손실

optimal_fraction = calculate_asymmetric_kelly(p_custom, w_custom, l_custom)

print(f"승률: {p_custom:.0%}, 수익률: {w_custom:.0%}, 손실률: {l_custom:.0%}")
print(f"최적의 베팅 비율(켈리 비율): {optimal_fraction:.2%}")
# 만약 f* = 4.0 이라면 400.00% 로 출력됩니다.
```

```python
print(f"{calculate_asymmetric_kelly(0.40, 0.5, 0.25):.2%}")     # 최적 자본 대비 투자비율  40%
```

```python
df = pd.read_csv("Data/Stocks/" + os.listdir("Data/Stocks")[3])
df["Date"] = pd.to_datetime(df["Date"])
df['MA120'] = df['Close'].rolling(window=120).mean()
df['MA60'] = df['Close'].rolling(window=60).mean()
df['MA20'] = df['Close'].rolling(window=20).mean()
df['MA5'] = df['Close'].rolling(window=5).mean()
df = df.set_index("Date")

def calculate_rsi(data, window = 14):
    delta = data.diff()
    gain = delta.where(delta > 0, 0).rolling(window = 14).mean()
    loss = -delta.where(delta < 0, 0).rolling(window = 14).mean()
    RS = gain / (loss + 1e-10)
    RSI = 100 - (100 / (1 + RS))
    return RSI
df["RSI"] = calculate_rsi(df["Close"])

df = df[((df["MA120"] < df["MA60"]) & (df["MA60"] < df["MA20"]) & (df["MA20"] < df["MA5"])) == True]
df = df[(df.index == "2025-07-02") & (df["Close"] > df["MA5"]) & (df["RSI"] >= 70)]
len(df)
```

```python
df = pd.read_csv("Data/Stocks/" + "현대차.csv")
df["Date"] = pd.to_datetime(df["Date"])
df = df.set_index("Date")
df
```

```python
from backtesting import Backtest, Strategy
from backtesting.lib import crossover
import pandas as pd
from backtesting import Backtest, Strategy
from backtesting.lib import crossover

# rsi 계산 함수

def rsi_func(series, n):   # 시리즈값, 며칠씩 묶을지 n값 
    s = pd.Series(series)
    delta = s.diff()
    gain = delta.where(delta > 0, 0).rolling(window = 14).mean()
    loss = -delta.where(delta < 0, 0).rolling(window = 14).mean()
    RS = gain / (loss + 1e-10)
    RSI = 100 - (100 / (1 + RS))
    return RSI

class RsiStrategy(Strategy):
    
    period = 14
    stock_buy = 70
    stock_sell = 30

    def init(self):
        self.rsi = self.I(rsi_func, self.data.Close, self.period)      # rsi값으로 초기값 설명

    def next(self):
        # RSI가 30을 상향 돌파하고, 현재 포지션이 없는 경우
        if crossover(self.rsi, self.stock_buy) and not self.position:
            self.buy()

        # RSI가 70을 하향 돌파하고 현재 포지션이 있을 때
        elif crossover(self.stock_sell, self.rsi) and self.position:
            self.position.close()
```

```python
bt = Backtest(df, RsiStrategy, cash = 1000000)

stats = bt.run()
stats

# 이전략 취하면 -3.33%, 그냥 묵혀두면 8.10%
```
