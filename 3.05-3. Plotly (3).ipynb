```python
# 애니메이션 
import matplotlib.pyplot as plt
import imageio.v2 as imageio
import os
import numpy as np

# 사인함수 그리기 
frames = np.linspace(0, np.pi * 2, 60)

box = []
for n, i in enumerate(frames):         # enumerate : 프레임값이 인덱스값과 같이 나오는 것 
    fig, ax = plt.subplots()
    x = np.linspace(0, np.pi * 2, 100)      # x축 100개 만듦 
    y = np.sin(x + i)     # 조금씩 이동하는 걸 보여주기 위해 프레임(i) 증가시킨다 
    ax.plot(x, y)
    plt.savefig(f"frame_{n}.png")
    box.append(f"frame_{n}.png")          # 프레임 1~60을 박스에 저장  (박스에 이미지 파일의 이름이 저장됨) 
    plt.show()        # show까지 해줘야 이미지 저장됨
```

```python
images = []
for i in box:
    images.append(imageio.imread(i))               # imread : 이미지를 읽어들여 

imageio.mimsave("sine.gif", images, fps = 30)       # mimsave : gif로 만들 수 있음
```

```python
for i in box:
    os.remove(i)          # 사진은 다 삭제
```
