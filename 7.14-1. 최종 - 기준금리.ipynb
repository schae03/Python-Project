```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 소비 → 실적
X = monthly_avg[['신용카드사용금액', '체크카드사용금액', '인터넷뱅킹거래금액', '스마트뱅킹거래금액']].dropna()

monthly_avg = monthly_avg.copy()
monthly_avg['월'] = monthly_avg.index.month  # 혹은 quarter

X['월'] = monthly_avg['월']

y = monthly_avg[['외환_수출실적금액', '외환_수입실적금액', '요구불입금금액']].dropna()
X, y = X.align(y, join='inner', axis=0)

# 시계열 순서 기반 분할
n = int(len(X) * 0.7)
X_train, X_test = X.iloc[:n], X.iloc[n:]
y_train, y_test = y.iloc[:n], y.iloc[n:]

# 모델 학습
# 트리 수 줄이기 + 최대 깊이 제한
rf_model1 = RandomForestRegressor(
    n_estimators=300,
    max_depth=5,
    min_samples_leaf=2,
    random_state=42
)
rf_model1.fit(X_train, y_train)

# 평가
print("Train R²:", rf_model1.score(X_train, y_train))
print("Test R²:", rf_model1.score(X_test, y_test))

# 중요도 시각화
importance1 = pd.Series(rf_model1.feature_importances_, index=X.columns)
plt.figure(figsize=(8, 4))
sns.barplot(x=importance1.values, y=importance1.index)
plt.title("소비 → 실적 변수 중요도")
plt.show()
```

```python
# 2. 실적 → 리스크
X2 = monthly_avg[['외환_수출실적금액', '외환_수입실적금액', '요구불입금금액']].dropna()
y2 = monthly_avg['고객등급점수'].dropna()
X2, y2 = X2.align(y2, join='inner', axis=0)

# 시계열 순서 기반 분할
n2 = int(len(X2) * 0.7)
X2_train, X2_test = X2.iloc[:n2], X2.iloc[n2:]
y2_train, y2_test = y2.iloc[:n2], y2.iloc[n2:]

# 모델 학습
rf_model2 = RandomForestRegressor(n_estimators=100, random_state=0)
rf_model2.fit(X2_train, y2_train)

# 평가
print("실적→리스크 RF R² (train):", rf_model2.score(X2_train, y2_train))
print("실적→리스크 RF R² (test):", rf_model2.score(X2_test, y2_test))

# 중요도 시각화
importance2 = pd.Series(rf_model2.feature_importances_, index=X2.columns)
plt.figure(figsize=(8, 4))
sns.barplot(x=importance2.values, y=importance2.index)
plt.title("실적 → 리스크 변수 중요도")
plt.show()
```

1. 구조방정식모형 (SEM)

```python
# !pip install semopy

from semopy import Model
import pandas as pd

# 등급 수치화
grade_map = {'최우수': 0, '우수': 1, '일반': 2}
df['고객등급점수'] = df['법인_고객등급'].map(grade_map)

# 월별 집계 (시계열 순서 유지)
monthly_avg = df.groupby('기준년월').agg({
    '여신_운전자금대출잔액': 'mean',
    '여신_시설자금대출잔액': 'mean',
    '신용카드사용금액': 'mean',
    '체크카드사용금액': 'mean',
    '인터넷뱅킹거래금액': 'mean',
    '스마트뱅킹거래금액': 'mean',
    '자동이체금액': 'mean',
    '외환_수출실적금액': 'mean',
    '외환_수입실적금액': 'mean',
    '요구불입금금액': 'mean',
    '고객등급점수': 'mean'
}).dropna().sort_index()

# SEM용 파생변수 생성
sem_df = pd.DataFrame(index=monthly_avg.index)
sem_df['대출'] = monthly_avg['여신_운전자금대출잔액'] + monthly_avg['여신_시설자금대출잔액']
sem_df['소비'] = (
    monthly_avg['신용카드사용금액'] + monthly_avg['체크카드사용금액'] +
    monthly_avg['인터넷뱅킹거래금액'] + monthly_avg['스마트뱅킹거래금액'] +
    monthly_avg['자동이체금액']
)
sem_df['실적'] = (
    monthly_avg['외환_수출실적금액'] + monthly_avg['외환_수입실적금액'] +
    monthly_avg['요구불입금금액']
)
sem_df['리스크'] = monthly_avg['고객등급점수']

# 표준화 (선택적으로 필요)
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
sem_std = pd.DataFrame(scaler.fit_transform(sem_df), columns=sem_df.columns, index=sem_df.index)

sem_std.head()
```

```python
from semopy import Model

desc = """
소비 ~ 대출
실적 ~ 소비
리스크 ~ 실적
"""

model = Model(desc)
model.fit(sem_std)
res = model.inspect()
print(res)
```

```python
from semopy import semplot
import matplotlib.pyplot as plt

fig = semplot(model, plot_ests=True)
plt.show()
```

2. Granger 인과관계 검정 (시계열 인과성 확인)

```python
from statsmodels.tsa.stattools import grangercausalitytests

# 예시: 소비 → 실적
data_gc = monthly_avg[['외환_수출실적금액', '신용카드사용금액']].dropna()
grangercausalitytests(data_gc, maxlag=3)
```

3. 시각 중심의 관계 탐색 (Time-Lag Cross-Correlation Plot)

```python
def cross_corr(x, y, lag=6):
    lags = range(-lag, lag+1)
    corrs = [x.corr(y.shift(l)) for l in lags]
    return pd.Series(corrs, index=lags)

cross_corr(monthly_avg['신용카드사용금액'], monthly_avg['외환_수출실적금액']).plot(kind='bar')
```
