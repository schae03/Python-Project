```python
# 데이터 집단화
import pandas as pd
import numpy as np

df = pd.read_csv("Data/Gapminder.tsv", sep = "\t")
A = df.groupby("year")["lifeExp"].mean()        # year열을 기준으로 집단화하겠다 (1952년은 1952년끼리, ...)
                      # lifeExp에 접근하겠다 / 평균값을 보겠다
A.plot()
```

```python
A = df.groupby(["year", "continent"])["lifeExp"].mean()        # 같은 연도내에서 대륙별로 집단화한 lifeExp의 평균 
A

A = df.groupby(["year", "continent"])["lifeExp"].describe()        # 같은 연도내에서 대륙별로 집단화한 lifeExp의 개수, 평균, 표준편차, ...
A

# A.to_excel("Result.xlsx")           # 이런 엑셀파일을 생성하겠다. 
```

```python
# 연도별로 기대수명이 가장 높은 나라 구하는 법
idx = df.groupby("year")["lifeExp"].idxmax()          # idxmax : 가장큰값이 있는 행의 위치값 꺼내줌 
df.loc[idx]          # idx번째 행을 찾아서 loc으로 접근
```

```python
# 대륙별로 국가가 몇개있을까?
# count(): 중복을 허용
# nunique(): 중복을 허용X     엔유니크
df.groupby("continent")["country"].nunique()
```

```python
# 열마다 다른 집계함수 적용 - agg 함수
df.groupby("과일").agg({"수량" : ["sum", "mean"],
                      "가격" : ["min", "max"]})
```

```python
import warnings
warnings.filterwarnings("ignore")

# 사용자 정의 함수 적용하기.
def func(x):
    return x[["수량","가격"]].sum()     # x라는 데이터프레임의  수량과 가격의 총합을 return하는 함수 
    
df.groupby("과일").apply(func)        # 만든함수 직접 적용할때에는 apply함수 사용 
```

```python
def func(x):
    A = x.sort_values("수량", ascending = False)       # 수량을 기준으로 내림차순 정렬
    return A[A["수량"] > 20]                 # 수량이 20개가 넘는 데이터프레임에 대해서만 합쳐나간다 

df.groupby("과일").apply(func)
```

```python
def func(x):
    new = pd.DataFrame({               # 데이터프레임을 하나씩 꺼내서 내가원하는 형태로 합치기 가능
        "sum" : [x["수량"].sum()],
        "mean" : [x["가격"].mean()],
        "max" : [x["수량"].max()]
    })
    return new

A = df.groupby("과일").apply(func)

A = A.reset_index()       # 여기서부턴선택사항 
del A['level_1']
A
```

```python
df["수량평균"] = df.groupby("과일")["수량"].transform("mean")        
# 과일별로 집단화하고 수량을 보고싶다. 근데 이걸 형태를 깨지 않고 옆에 붙여서 보고싶음   =>   transform
df
```

```python
# 원하는 그룹만 필터링
def func(x):
    return x["수량"].sum() >= 40

df.groupby("과일").filter(func)         # 내가원하는 그룹만 필터링하겠다    =>    filter
```
