# EDA

```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
import koreanize_matplotlib
import seaborn as sns

# 시각화 설정
sns.set(style="whitegrid")  
plt.rcParams['font.family'] = 'Malgun Gothic'       # 한글이면 'Malgun Gothic' 또는 'AppleGothic'

# -------------------------
# 1. member 테이블 EDA
# -------------------------

# 성별 분포
sns.countplot(data=member, x='성별')
plt.title('성별 분포')
plt.show()

# 연령 분포
sns.histplot(member['연령'], bins=10, kde=True)
plt.title('연령 분포')
plt.show()

# 결혼유무 vs 자녀여부
sns.countplot(data=member, x='결혼유무', hue='자녀여부')
plt.title('결혼유무와 자녀여부 관계')
plt.show()

# 유입경로 분포
member['유입경로'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('유입경로 비율')
plt.ylabel('')
plt.show()
```

```python
# -------------------------
# 2. product 테이블 EDA
# -------------------------

# 물품대분류별 상품 수
product['물품대분류'].value_counts().plot(kind='bar')
plt.title('물품대분류별 상품 수')
plt.show()

# 제품단가 분포
sns.boxplot(x='물품대분류', y='제품단가', data=product)
plt.xticks(rotation=45)
plt.title('제품 단가 분포 by 대분류')
plt.show()

# 판매량 상위 상품
top_products = product.sort_values(by='물품판매량', ascending=False).head(10)
sns.barplot(data=top_products, x='상품명', y='물품판매량')
plt.title('판매량 상위 10개 상품')
plt.xticks(rotation=90)
plt.show()
```

```python
# -------------------------
# 3. sales 테이블 EDA
# -------------------------

# 구매일 → datetime 변환
sales['구매일'] = pd.to_datetime(sales['구매일'])

# 월별 구매금액 추이
sales['월'] = sales['구매일'].dt.to_period('M').astype(str)
monthly_sales = sales.groupby('월')['구매금액'].sum().reset_index()
sns.lineplot(data=monthly_sales, x='월', y='구매금액')
plt.xticks(rotation=45)
plt.title('월별 구매금액 추이')
plt.show()

# 고객별 총구매액 상위 10명
top_customers = sales.groupby('고객번호')['구매금액'].sum().sort_values(ascending=False).head(10)
top_customers.plot(kind='bar')
plt.title('구매금액 상위 고객')
plt.ylabel('총 구매금액')
plt.show()

# 결제금액 vs 구매금액 비교
sns.scatterplot(data=sales, x='구매금액', y='결제금액')
plt.title('구매금액 vs 결제금액')
plt.show()

# 배송 소요일 계산
sales['배송소요일'] = (pd.to_datetime(sales['배송완료일']) - pd.to_datetime(sales['배송시작일'])).dt.days
sns.histplot(sales['배송소요일'], bins=20, kde=True)
plt.title('배송 소요일 분포')
plt.show()
```

```python
# -------------------------
# 4. 통합 분석 예시 (member + sales)
# -------------------------

merged = sales.merge(member, on='고객번호', how='left')

# 성별별 평균 구매금액
sns.barplot(data=merged, x='성별', y='구매금액')
plt.title('성별별 평균 구매금액')
plt.show()

# 연령대 구간화 후 시각화
merged['연령대'] = pd.cut(merged['연령'], bins=[20, 30, 40, 50, 60], labels=['20대', '30대', '40대', '50대'])
sns.boxplot(data=merged, x='연령대', y='구매금액')
plt.title('연령대별 구매금액 분포')
plt.show()
```

## 성별이 남자인 경우의 평균 구매금액 및 결제금액

```python
sns.countplot(data=member, x='성별')
plt.title('성별 분포')
plt.show()
```

```python
# 성별별 평균 구매금액 및 평균 결제금액 집계
gender_mean = merged.groupby('성별')[['구매금액', '결제금액']].mean().reset_index()

# 결과 출력
print(gender_mean)
```

```python
# 막대그래프로 비교
gender_mean_melted = gender_mean.melt(id_vars='성별', value_vars=['구매금액', '결제금액'],
                                             var_name='구분', value_name='금액')

sns.barplot(data=gender_mean_melted, x='성별', y='금액', hue='구분')
plt.title('성별별 평균 구매금액 vs 평균 결제금액')
plt.ylabel('금액 (원)')
plt.xlabel('성별')
plt.tight_layout()
plt.show()
```

### ㄴ 남자가 여자보다 구매 건수는 적지만, 한번살때 더 많은 금액을 결제한다

```python
merged[merged["성별"] == "남"].groupby("물품대분류").size().sort_values(ascending = False)
```

```python
# 제품단가가 0이 아닌 상품만 필터링
valid_product = product[product["제품단가"] > 0]

# 물품대분류별 평균 제품단가 계산
avg_price_by_category = (
    valid_product.groupby("물품대분류")["제품단가"]
    .mean()
    .sort_values(ascending = False)
    .reset_index()
    .rename(columns={"제품단가": "평균제품단가"})
)

# 결과 확인
avg_price_by_category
```

ㄴ 남자가 많이 사는 제품대분류가 실제로 평균 제품단가가 높지는 않음   
제품단가가 높아서 결제금액이 높은 건 아닐수도
