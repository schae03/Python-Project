```python
import pandas as pd
df = pd.read_csv("Data/딥러닝 프로젝트/대구광역시_카드금액sum.csv")
df
```

```python
# 거래년월이 문자열 또는 정수형인 경우 처리
df[['year', 'month']] = df['거래연월'].str.split('-', expand=True)
df['year'] = df['year'].astype(int)
df['month'] = df['month'].astype(int)
df['date'] = pd.to_datetime(df[['year', 'month']].assign(day=1))
df
```

```python
df2 =  df[df["가맹점업종명"].isin([    '일반한식', '일식·회집', '중국식', '서양음식', '한정식', '정장', '주점', '단란주점', '유흥주점', '칵테일바',
    '노래방', '당구장', '볼링장', '골프연습장', '스크린골프', '영화관', '종합레져타운', '레져업소', '사우나',
    '화장품', '기타대인서비스', '피부미용실', '안마/스포츠마사지', '헬스클럽', '스넥', '카인테리어', '렌터카',
    '기념품점', '화원', '완구점', '애완동물', '문화취미기타', '민예·공예품', '음반영상물', '티켓', '관광여행',
    '화방·표구점',    '정장', '캐주얼의류', '아동의류', '스포츠의류', '내의판매', '맞춤복점',
    '액세서리', '귀금속', '시계', '기타의류', '잡화', '기타잡화',
    '인터넷Mall', '인터넷종합Mall'])]
df2
```

```python
pd.DataFrame(df2.groupby(["year", "month"])["승인금액"].mean())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib

# groupby로 연월별 승인금액 평균 계산
df_grouped = df.groupby(["year", "month"])["승인금액"].mean().reset_index()

# 연월을 datetime 형식으로 변환
df_grouped["date"] = pd.to_datetime(df_grouped["year"].astype(str) + "-" + df_grouped["month"].astype(str).str.zfill(2))

# 그래프 그리기
plt.figure(figsize=(12, 6))
plt.plot(df_grouped["date"], df_grouped["승인금액"], marker='o', label="월별 평균 승인금액")

plt.title("연월별 평균 승인금액 추이")
plt.xlabel("기준 연월")
plt.ylabel("평균 승인금액")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib

# groupby로 연월별 승인금액 총합 계산
df_grouped = df2.groupby(["year", "month"])["승인금액"].sum().reset_index()

# 연월을 datetime 형식으로 변환
df_grouped["date"] = pd.to_datetime(df_grouped["year"].astype(str) + "-" + df_grouped["month"].astype(str).str.zfill(2))

# 그래프 그리기
plt.figure(figsize=(12, 6))
plt.plot(df_grouped["date"], df_grouped["승인금액"], marker='o', label="월별 평균 승인금액")

plt.title("연월별 승인금액의 총합 추이 (쾌락적소비)")
plt.xlabel("기준 연월")
plt.ylabel("총 승인금액")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
