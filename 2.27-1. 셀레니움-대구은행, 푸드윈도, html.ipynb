# 대구은행 검색했을때 나오는 뉴스기사 가져오기

```python
## 해당 페이지의  뉴스 기사의 제목, 내용만 가져와라 

#내가한거
box = []
for i in range(10):
    title = driver.find_elements(By.CLASS_NAME, "news_tit")[i].text
    content = driver.find_elements(By.CLASS_NAME, "news_dsc")[i].text
    box.append([title, content])
```

```python
#강사님
box = []

news = driver.find_element(By.CSS_SELECTOR, "ul.list_news._infinite_list").find_elements(By.CSS_SELECTOR, "li.bx")
# 뉴스기사 하나의 제목,내용 등 다포함하는 항목에 대해          해당 페이지 내에서 그런걸 다찾아라
for i in news:
    title = (i.find_element(By.CSS_SELECTOR, "a.news_tit").get_attribute("title"))
            # 해당 뉴스기사에서 news_tit 찾아서 타이틀이라는 속성을 가져와라 
    summary = (i.find_element(By.CSS_SELECTOR, "div.dsc_wrap").text.strip())
    A = [title, summary]
    if A not in box:
        box.append(A)          # 같은 내용 중복해서 들어가지 않게  기존에 없는 것만 추가 

box
```

```python
# 페이지 스크롤하는 방법 1 - 확 내림 

driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")  
# 스크롤의 바디부분만큼 내려주겠다 
```

```python
# 페이지 스크롤하는 방법 2 - 조금씩 

body = driver.find_element(By.TAG_NAME, "body")
for n in range(10):
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(2)
```

```python
# 뉴스기사 100개의 제목, 본문요약 수집하겠다 

box = []
while True:
    news = driver.find_element(By.CSS_SELECTOR, "ul.list_news._infinite_list").find_elements(By.CSS_SELECTOR, "li.bx")
    for i in news:
        title = (i.find_element(By.CSS_SELECTOR, "a.news_tit").get_attribute('title'))
        summary = (i.find_element(By.CSS_SELECTOR, "div.dsc_wrap").text.strip())
        A = [title, summary]
        if A not in box:
            box.append(A)
            if len(box) == 100:
                break
    if len(box) == 100:
        break
    print(f"{len(box)}개 뉴스기사 수집 완료")
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    # 가져올 거 다 가져오고 스크롤 내리겠다는 뜻
    time.sleep(3)
```

```python
len(box)
```

```python
import pandas as pd
df = pd.DataFrame(box, columns = ["제목", "내용"])
df.to_excel("네이버 뉴스기사.xlsx", index = False)
df
```

# 네이버스토어 푸드윈도의 가격,제목,별점,리뷰개수 100개

```python
driver = uc.Chrome()
driver.get("https://shopping.naver.com/fresh/directfarm/home")
```

```python
# 내가한거 
box = []
while True:
    items = driver.find_element(By.CSS_SELECTOR, "ul._2apNeE5N3I").find_elements(By.CSS_SELECTOR, "li._1OKV3UfxuI")
    for i in items:
        title = (i.find_element(By.CSS_SELECTOR, "strong._1lxNNH5gfD").text.strip())
        price = (i.find_element(By.CSS_SELECTOR, "span._3MzrqSUlwR").text.strip())
        star = (i.find_element(By.CSS_SELECTOR, "li._3G5BbVc9PL._1FOtzXHiH3").text.strip("평점").strip())
        review = (i.find_element(By.CSS_SELECTOR, "li._3G5BbVc9PL.a0tLQiDclf").text.strip("리뷰"))
        A = [title, price, star, review]
        if A not in box:
                box.append(A)
                if len(box) == 100:
                    break
    if len(box) == 100:
        break
    print(f"{len(box)}개 상품 수집 완료")
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(3)
```

```python
import pandas as pd
df = pd.DataFrame(box, columns = ["이름","가격","별점","리뷰개수"])
df.to_excel("네이버스토어 푸드윈도.xlsx", index = False)
df
```

```python
# 강사님 
box = []
while True:
    for i in driver.find_elements(By.CSS_SELECTOR, "li._1OKV3UfxuI"):
        title = (i.find_element(By.CSS_SELECTOR, "strong._1lxNNH5gfD").text.strip())
        price = (i.find_element(By.CSS_SELECTOR, "span._3MzrqSUlwR").text)
        star = (i.find_element(By.CSS_SELECTOR, "span._29gNnX1yLx").text.split("\n")[1])
        review = (i.find_elements(By.CSS_SELECTOR, "span._29gNnX1yLx")[1].text)
        A = [title, price, star, review]
        if A not in box:
            box.append(A)
            if len(box) == 100:
                break
    print(f"{len(box)}개 아이템 수집 완료")
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(3)
    if len(box) == 100:
        break
```

```python
# 상품에 있는 이미지를 가져오고 싶다
import requests

idx = 0
for i in driver.find_elements(By.CSS_SELECTOR, "li._1OKV3UfxuI"):
    image = (i.find_element(By.TAG_NAME, "img")) 
    src = (image.get_attribute("src"))       # src 속성 쓰면 이미지를 링크로 가져와줌
    if not src:
        continue          # src에 속성 없으면 패스하도록
    for j in [".jpg", ".png", ".jpeg", ".gif"]:
        if j in src:
            if j in src:
                f = open(f"{idx}{j}", "wb")    # 이미지 파일로 하려면 binary 형태로. 즉 wb
                respone = requests.get(src)
                f.write(respone.content)
                f.close()
                idx += 1
```

```python
from bs4 import BeautifulSoup
driver.get("https://finance.naver.com/item/sise_day.naver?code=005930")
```

```python
import warnings
warnings.filterwarnings("ignore")

html = BeautifulSoup(driver.page_source)     # 현재보고있는페이지의 소스코드를 받아와라 
table = html.find("table")
table = pd.read_html(str(table))[0].dropna()
table
```

```python
# 1~10페이지 테이블을 셀레니움으로 가져와서 하나의 데이터프레임 만들기
box = []
for i in range(1, 11):
    driver.get(f"https://finance.naver.com/item/sise_day.naver?code=005930&page={i}")
    time.sleep(2)
    html = BeautifulSoup(driver.page_source)     # 현재보고있는페이지의 소스코드를 받아와라 
    table = html.find("table")
    table = pd.read_html(str(table))[0].dropna()
    box.append(table)
```

```python
pd.concat(box)
```

```python
# 모든 페이지의 테이블을 셀레니움으로 가져와서 하나의 데이터프레임 만들기
box = []
for i in range(1, 720):
    driver.implicitly_wait(10)         # 페이지가 모두 뜰때(HTML한정)까지 최대 10초를 기다려라 
    driver.get(f"https://finance.naver.com/item/sise_day.naver?code=005930&page={i}")
    html = BeautifulSoup(driver.page_source)     # 현재보고있는페이지의 소스코드를 받아와라 
    table = html.find("table")
    table = pd.read_html(str(table))[0].dropna()
    box.append(table)
```

```python
samsung = pd.concat(box, ignore_index=True)
samsung = samsung.iloc[1:]
samsung.to_excel("삼성 주가 데이터.xlsx", index = False)
samsung
```
