```python
df = pd.read_excel("Data/text data.xlsx")       # 텍스트 데이터
del df["label"]

knu = pd.read_csv("Data/knu_sentiment_lexicon.csv")     # 감성 사전  -  좋은건 1또는2, 안좋은건 -1또는-2
```

```python
from tqdm import tqdm

box = []
for i in tqdm(df["document"]):
    i = str(i)
    score = 0
    for w, p in zip(knu["word"], knu["polarity"]):
        if w in i:      # 감성사전에 있는게 리뷰에 있으면
            score += (p * i.count(w))
    box.append(score)
```

```python
box
```

```python
box2 = []
for i in box:
    if i > 0:
        box2.append("긍정")
    elif i < 0:
        box2.append("부정")
    else:
        box2.append("중립")

df["감성분석결과"] = box2
df
```

## RNN, LSTM : 기억력 문제 // 트랜스포머(셀프 어텐션)
It is a cute animal.

```python
import torch
import torch.nn.functional as F

# "it", "is", "a", "cute", "animal"
sentence_vectors = torch.tensor([
    [1.0, 2.0, 0.1],  # it
    [0.1, 0.5, 2.1],  # is
    [0.2, 2.2, 0.2],  # a
    [2.3, 0.3, 0.3],  # cute
    [1.1, 1.9, 0.2]   # animal
])
```

```python
A = torch.matmul(sentence_vectors[0], sentence_vectors.T)
F.softmax(A, dim = 0)
```
