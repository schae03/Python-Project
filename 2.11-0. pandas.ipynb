```python
# 시리즈 연산
s1 = pd.Series([1,2,3])
s2 = pd.Series([1,2,3], index = [2,0,1])

s1 + s2        # 같은 행이름 가진 값끼리 연산. (같은 위치에 있는 값끼리 더해진 것 X)
```

```python
import numpy as np

s1 = pd.Series([1,2,3])
s2 = pd.Series([1,np.nan,3])       # np.nan : 누락값

s1 + s2       # 어떤값 + 누락값 => 누락값
s1.add(s2, fill_value=0)      # add : 누락값이 있는경우에  0으로 채우고싶을때
```

```python
np.sqrt(s1)
```

```python
# 특정 조건을 만족하는 행만 추출하기.
# 불 인덱싱 

A = df['lifeExp'] >= 50 
df[A]               # 위의 조건에서 true인 행만 가져옴 (기대수명이 50 이상)
df[~A]                    # false인 행만  (기대수명이 50 미만인 행만)
```

```python
# 아프리카이면서 기대수명이 30 이하인 행만 추출하기
A = df['continent'] == 'Africa'
B = df['lifeExp'] <= 30

df[A & B]
```

```python
# 퀴즈: 기대수명이 50이상이면서 국가가 Japan인 행만 추출하기
A = df['lifeExp'] >= 50
B = df['country'] == 'Japan'
df[A & B]

# 한줄로 적으려면 괄호로 묶어주기
df[(df['lifeExp'] >= 50) & (df['country'] == 'Japan')]
```

```python
# 대륙이 아시아, 아프리카, 유럽인 행만 가져오기

df['continent'].isin(['Africa','Asia','Europe'])        # isin: 조건 여러개 만들 필요없이 여러개의 행 추출 가능. 
df[A]
```

```python
# lifeExp가 20~30까지인 행만 추출하기.
A = df['lifeExp'].between(20, 30)              # between: 두 범위 사이에 있는 행 추출 가능
df[A]
```

```python
# 특정 문자가 포함된 행만 가져오기.
df[df['continent'].str.contains("n")]            # 대륙열에서 소문자 N이 포함된 열만 가져온다.
df[df['continent'].str.contains("n", case = False)]         # 대소문자 구분없이 n(N)이 포함된 열 가져온다.
```

```python
df =  pd.read_excel("Data/성적 처리.xlsx", sheet_name = "Sheet1")
df
```

```python
# 정렬과 순위

# axis가 0이면 아래방향, 1이면 오른쪽방향
df["총점"] = df.iloc[:, 2:].sum(axis = 1)       # 기본은 아래방향 (axis = 0).  오른쪽방향으로 더하려면 axis = 1
df["평균"] = df["총점"] / 5
df
```

```python
df.sort_values("평균")       # 정렬하고싶은 열을 기준으로 오름차순으로 정렬    //    내림차순으로 표현하려면 ascending = False
df.sort_values(["반", "평균"], ascending = [True, False])      # 반은 오름차순으로, 각 반에 대해 평균은 내림차순으로
```

```python
# 행 정렬  :  sort_index
df.sort_index(ascending = False)
```

```python
# 문자열 / 사전순으로 정렬(가나다순)  :  sort_index
df.sort_index(axis = 1)
```

```python
# 순위 구하기  :  rank

# (1) 3등이 2명이면 둘다 3.5등을 처리하고 다음 등수롤 5등으로 (3, 4등없음)
df["순위"] = df["평균"].rank(ascending = False, method = "average")      
df
```

```python
# (2) 3등이 2명이면 둘다 4등으로 처리하고 다음 등수롤 5등으로 (3등없음)
df["순위"] = df["평균"].rank(ascending = False, method = "max")      
df
```

```python
# (3) 3등이 2명이면 둘다 3등으로 처리하고 다음 등수롤 5등으로 (4등없음)
df["순위"] = df["평균"].rank(ascending = False, method = "min")     
df
```

```python
# (4) 3등이 2명이면 먼저 발견한 3등이 3등, 다음 발견된 3등은 4등으로 처리.
df["순위"] = df["평균"].rank(ascending = False, method = "first")     
df
```

```python
# (5) min과 같다.   3등이 2명이면 둘다 3등으로 처리하고 다음 등수롤 4등으로. (건너뛰지X)
df["순위"] = df["평균"].rank(ascending = False, method = "dense")     
df
```

```python
df["순위"] = df["평균"].rank(ascending = False, 
                         method = "dense", pct = True)       # pct = True  :  백분위 순위로 바꿔서 구할 수 O
df
```
