6. 각 ```자재명``` 별로 ```쿠팡출고수량``` 과 ```입고수량``` 의 평균을 계산한 후,
- 계산된 결과에서 **쿠팡출고수량** 평균을 기준으로
    - 상위 10%에 해당하는 자재명을 'A',
    - 상위 30%를 'B',
    - 상위 60%를 'C',
    - 나머지를 'D'로 분류하는 **등급분류** 파생 변수를 생성하고,
- 각 등급별로 가장 빈번하게 등장하는 카테고리 항목을 확인하시오.

````python
# 자재별 평균 계산
material_avg = df.groupby('자재명').agg({'쿠팡출고수량': 'mean', '입고수량': 'mean'}).reset_index()
material_avg
````

````python
# 평균 쿠팡출고수량 기준 내림차순 정렬
material_avg = material_avg.sort_values('쿠팡출고수량', ascending=False).reset_index(drop=True)
n = len(material_avg)

# 등급 부여 (인덱스 기준 비율 적용)
material_avg['등급'] = 'D'  # 기본 등급 D
material_avg.loc[:int(n*0.1)-1, '등급'] = 'A'
material_avg.loc[int(n*0.1):int(n*0.3)-1, '등급'] = 'B'
material_avg.loc[int(n*0.3):int(n*0.6)-1, '등급'] = 'C'

print("자재별 평균과 등급 분류 (상위 일부만 확인):")
material_avg
````

````python
# 원본 데이터에 등급 정보를 병합하여 각 등급별 카테고리 빈도수 확인
df_material = pd.merge(df, material_avg[['자재명', '등급']], on='자재명', how='left')
grade_categories = df_material.groupby('등급')['카테고리'].agg(lambda x: x.value_counts().index[0])
print("각 등급별 가장 빈번하게 등장하는 카테고리:")
print(grade_categories)
````

7. 사용자로부터 특정 ```카테고리```를 입력받은 후, 해당 카테고리의 데이터에서
    - ```발주가능상태```별 및 날짜 컬럼을 활용해 계산한 ```월```별 ```쿠팡출고수량``` 합계를 산출하고 시각화하며,
    - 결과 그래프를 `카테고리명_월별_쿠팡출고량.png` 파일로 저장하는 함수를 작성하시오.

````python
import matplotlib.pyplot as plt
import koreanize_matplotlib

def visualize_monthly_outbound(category):
    # 해당 카테고리 데이터 필터링
    cat_data = df[df['카테고리'] == category].copy()
    
    # 날짜가 datetime 형식인지 확인 (이미 처리된 경우 생략 가능)
    cat_data['날짜'] = pd.to_datetime(cat_data['날짜'])
    
    # 월별 집계를 위해 'month' 컬럼(연-월 형태) 생성
    cat_data['month_period'] = cat_data['날짜'].dt.to_period('M')
    monthly_sum = cat_data.groupby('month_period')['쿠팡출고수량'].sum().reset_index()
    monthly_sum['month_period'] = monthly_sum['month_period'].astype(str)
    
    # 시각화 (라인 플롯)
    plt.figure(figsize=(10, 6))
    plt.plot(monthly_sum['month_period'], monthly_sum['쿠팡출고수량'], marker='o')
    plt.title(f'{category} - 월별 쿠팡출고수량')
    plt.xlabel('월')
    plt.ylabel('쿠팡출고수량 합계')
    plt.grid(True)
    plt.xticks(rotation=45)
    
    # 파일로 저장
    filename = f"{category}_월별_쿠팡출고량.png"
    plt.savefig(filename, bbox_inches='tight')
    plt.close()
    print(f"그래프가 '{filename}'로 저장되었습니다.")
````

````python
visualize_monthly_outbound("덴탈케어")
````

9. 각 카테고리별로 7월과 8월의 ```쿠팡출고수량```에 대한 평균 및 합계를 산출한 후,
    - 두 달 간의 차이를 계산하시오.

````python
# 7월과 8월 데이터 필터링 (2019년)
data_jul_aug = df[(df['year'] == 2019) & (df['month'].isin([7, 8]))].copy()

# 카테고리와 월별로 평균과 합계 계산
grouped = data_jul_aug.groupby(['카테고리', 'month'])['쿠팡출고수량'].agg(['mean', 'sum']).reset_index()

# 피벗테이블 형태로 변환
pivot_mean = grouped.pivot(index='카테고리', columns='month', values='mean')
pivot_sum = grouped.pivot(index='카테고리', columns='month', values='sum')

# 7월과 8월 차이 계산 (8월 - 7월)
pivot_mean['diff'] = pivot_mean[8] - pivot_mean[7]
pivot_sum['diff'] = pivot_sum[8] - pivot_sum[7]

print("[7월과 8월] 카테고리별 쿠팡출고수량 평균 차이 (8월-7월):")
print(pivot_mean['diff'])
````

````python
print("[7월과 8월] 카테고리별 쿠팡출고수량 합계 차이 (8월-7월):")
print(pivot_sum['diff'])
````
