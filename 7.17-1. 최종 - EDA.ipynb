```python
import pandas as pd

df_reset = df.reset_index(drop=True)
df2_reset = df2.reset_index(drop=True)

concat = pd.DataFrame(
    [row for pair in zip(df_reset.values, df2_reset.values) for row in pair],
    columns=df.columns
)
concat
```

```python
concat["구분"].unique()
```

```python
# 1. 날짜 컬럼 추출
value_cols = [col for col in concat.columns if '년' in col]
essential_cols = ['금융회사', '구분'] + value_cols

# 2. 필요한 구분만 필터링
df_filtered = concat[essential_cols]
df_filtered = df_filtered[df_filtered['구분'].isin(['비이자수익', '영업수익'])]

# 3. pivot_table로 변환 (구분이 컬럼으로 올라오게 됨 → MultiIndex 생성됨)
df_pivot = df_filtered.pivot_table(
    index='금융회사',
    columns='구분',
    values=value_cols,
    aggfunc='sum'  # 혹시 동일한 회사+구분이 여러 행일 경우 대응
)

# 4. 비이자수익 / 영업수익 계산 (다중인덱스 처리)
# df_pivot의 columns → MultiIndex (value_cols, 구분)
# 예시: df_pivot.columns = MultiIndex([('2014년03월말', '비이자수익'), ('2014년03월말', '영업수익'), ...])

# 계산: 모든 기간에 대해 비이자수익 / 영업수익
비이자 = df_pivot.xs('비이자수익', axis=1, level=1)
영업   = df_pivot.xs('영업수익', axis=1, level=1)

# 비율 = 비이자.where(영업 > 0) / 영업.where(영업 > 0)
비율 = 비이자.where((영업 > 0) & (비이자 > 0)) / 영업.where((영업 > 0) & (비이자 > 0))
비율 = 비율.where(비율 > 0.01)

# 비율  = 비이자 / 영업

# 보기 좋게: 컬럼 이름에 비율 표시
비율.columns = [f"{col}_비이자비율" for col in 비율.columns]

# 결과
비율 = 비율.reset_index()
비율
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib  # 한글 깨짐 방지
import matplotlib.ticker as mtick

# wide → long 형태 변환
df_long = pd.melt(
    비율, 
    id_vars='금융회사', 
    var_name='기준월', 
    value_name='비이자수익비율'
)

# 날짜 문자열 정리
df_long['기준월'] = df_long['기준월'].str.replace('_비이자비율', '')

# 기업별 시계열 그래프
plt.figure(figsize=(16, 6))

# 전체 평균 계산 (기준월별 평균)
mean_df = df_long.groupby('기준월', as_index=False)['비이자수익비율'].mean()

# 기업별 선
for company in df_long['금융회사'].unique():
    subset = df_long[df_long['금융회사'] == company]
    plt.plot(subset['기준월'], subset['비이자수익비율'], marker='o', label=company, linewidth=1.5)

# 전체 평균선 (가장 굵게 강조)
plt.plot(mean_df['기준월'], mean_df['비이자수익비율'], 
         label='전체 평균', color='black', linewidth=5)

plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1.0))  # 비율이 0~1 사이일 경우

# 시각화 설정
plt.xticks(rotation=45)
plt.ylabel('비이자수익 비중')
plt.title('금융회사별 비이자수익비율 추이 및 전체 평균')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 1. 날짜 컬럼 추출
date_cols = [col for col in df.columns if '년' in col]

# 2. 관심 항목 지정
이자수익 = ['    이자수익']
비이자항목들 = [
    '    수수료수익', '        원화수입수수료', '        외화수입수수료', '        금수입수수료',
    '    기타영업수익',    '        기타포괄손익-공정가치측정유가증권관련수익',
    '        상각후원가측정유가증권관련수익', '        기타수익'
]
target_items = 이자수익 + 비이자항목들

# 3. 필요한 구분만 필터링
df_filtered = df[df['구분'].isin(target_items)]

# 4. 평균 계산: (금융회사, 구분) 기준으로 월별 평균
mean_df = df_filtered.groupby(['금융회사', '구분'])[date_cols].mean()

# 5. 항목별 평균값만 뽑기: 각 (금융회사, 구분)에 대해 전체 기간 평균
mean_df['평균'] = mean_df[date_cols].mean(axis=1)

# 6. 피벗: 행 → 금융회사, 열 → 구분, 값 → 평균값
pivot_avg = mean_df['평균'].unstack('구분')  # 결과: [금융회사 x 구분] 구조

# 7. 이자수익 vs 비이자 항목 상관계수 계산
corr_list = []
for col in pivot_avg.columns:
    if col == '    이자수익':
        continue
    corr = pivot_avg['    이자수익'].corr(pivot_avg[col])
    corr_list.append((col.strip(), round(corr, 2)))  # 공백 제거해서 예쁘게

# 8. 표로 출력
corr_df = pd.DataFrame(corr_list, columns=['항목', '상관계수'])
print("<표 1> 이자수익과 비이자수익 항목별 상관관계\n")
print(corr_df.to_string(index=False))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib

# 1. 기준금리: 일별 → 분기별 평균으로 변환
df['date'] = pd.to_datetime(df['date'])
df['분기'] = df['date'].dt.to_period('Q').dt.to_timestamp(how='end')  # '2014Q1' → '2014-03-31'

# 분기별 평균 기준금리 계산
rate_quarterly = df.groupby('분기')['rate'].mean().reset_index()
rate_quarterly.rename(columns={'분기': '기준월', 'rate': '기준금리'}, inplace=True)

# 2. NIM 데이터 전처리
nim_filtered = nim[nim['코드'] == 'C'].copy()

# '2014년03월말' 등 날짜형 컬럼 추출
nim_date_cols = [col for col in nim_filtered.columns if '년' in col]

# wide → long 변환
nim_long = nim_filtered.melt(
    id_vars=['금융회사'], 
    value_vars=nim_date_cols, 
    var_name='기준월', 
    value_name='NIM'
)

# 날짜 포맷 통일 (문자열 → datetime)
nim_long['기준월'] = pd.to_datetime(nim_long['기준월'].str.replace('년', '-').str.replace('월말', '').str.replace(' ', ''), errors='coerce')

# 3. 기준금리와 병합
merged = pd.merge(nim_long, rate_quarterly, on='기준월', how='left')

# 4. 시각화
plt.figure(figsize=(14, 6))

# NIM 라인 (은행별)
for bank in merged['금융회사'].unique():
    subset = merged[merged['금융회사'] == bank]
    plt.plot(subset['기준월'], subset['NIM'], marker='o', label=f'NIM - {bank}')

# 기준금리 라인
plt.plot(rate_quarterly['기준월'], rate_quarterly['기준금리'], color='black', linestyle='--', linewidth=2, label='기준금리')

# 마무리 시각화
plt.title('기준금리 vs IM뱅크·시중은행 NIM 추이')
plt.xlabel('기준월')
plt.ylabel('비율 (%)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# 1. 파일 불러오기
# 기준금리 데이터 (일 단위 → 월 단위 평균)
rate_df = pd.read_excel("Data/최종 프로젝트/기준금리.xlsx", parse_dates=['date'])
rate_df['연월'] = rate_df['date'].dt.to_period('M')
monthly_rate = rate_df.groupby('연월')['rate'].mean().reset_index()
monthly_rate['연월'] = monthly_rate['연월'].astype(str)

# NIM 데이터 불러오기
nim_df = nim.copy()

# 2. NIM 데이터 전처리
nim_df = nim_df[nim_df['코드'] == 'C'].copy()
nim_df = nim_df.drop(columns=['금융회사코드', '권역금융회사코드', '구분', '코드'])

# 열 정리: wide → long
nim_long = nim_df.melt(id_vars='금융회사', var_name='연월', value_name='NIM')
nim_long['연월'] = nim_long['연월'].str.replace('년', '-').str.replace('월말', '').str.strip()
nim_long = nim_long[nim_long['연월'].str.len() == 7]  # yyyy-mm 형태만

# 3. 기준금리와 병합
merged_df = pd.merge(nim_long, monthly_rate, how='left', on='연월')
merged_df = merged_df.dropna()

# 4. 수치 분석
results = []

for bank in merged_df['금융회사'].unique():
    sub = merged_df[merged_df['금융회사'] == bank]
    
    if len(sub) < 10:
        continue  # 데이터 부족 시 건너뜀
    
    # 상관계수
    corr = sub['NIM'].corr(sub['rate'])
    
    # 회귀분석: ΔNIM / Δ기준금리
    X = sub[['rate']]
    y = sub['NIM']
    model = LinearRegression().fit(X, y)
    coef = model.coef_[0]
    
    # 평균, 표준편차, 변동계수
    mean_val = y.mean()
    std_val = y.std()
    cv = std_val / mean_val if mean_val != 0 else np.nan
    
    results.append({
        '금융회사': bank,
        '상관계수': round(corr, 3),
        'ΔNIM/Δ기준금리': round(coef, 3),
        'NIM 평균': round(mean_val, 3),
        'NIM 표준편차': round(std_val, 3),
        '변동계수(CV)': round(cv, 3)
    })

# 5. 결과 출력
result_df = pd.DataFrame(results)
result_df
```

```python
# 원하는 금융회사 순서 리스트
ordered_banks = [
    '국민은행',
    '신한은행',
    '아이엠뱅크',
    '우리은행',
    '주식회사 하나은행',
    '한국스탠다드차타드은행',
    '한국씨티은행'
]

# 순서에 따라 카테고리 정렬 후 재정렬
result_df['금융회사'] = pd.Categorical(result_df['금융회사'], categories=ordered_banks, ordered=True)
result_df = result_df.sort_values('금융회사').reset_index(drop=True)
result_df
```
