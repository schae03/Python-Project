```python
member["고객번호"] = member["고객번호"].astype(str)
sales["고객번호"] = sales["고객번호"].astype(str)
```

```python
merged = sales.merge(member, on='고객번호', how='left')
merged["고객번호"] = merged["고객번호"].astype(str)
merged
```

```python
merged["성별"].isnull().sum()
```

```python
# 병합 결과에서 자녀여부가 NaN인 고객번호
missing_ids = merged[merged["자녀여부"].isnull()]["고객번호"].unique()

# 이 고객들이 member에 있는지 확인
missing_check = pd.Series(missing_ids).isin(member["고객번호"])

# False가 있으면 => member에 없는 고객이 병합되면서 NaN 발생한 것
print("member에 존재하는가?", missing_check.value_counts())
```

```python
null_child = merged[(merged["자녀여부"].isnull()) & (merged["구매시월령"].isnull())]

# 고객번호별 주문 건수 세기
order_counts = (
    null_child.groupby("고객번호")["주문번호"]
    .nunique()  # 주문번호가 중복될 수 있으므로 고유값 기준
    .reset_index()
    .rename(columns={"주문번호": "주문건수"})
    .sort_values(by="주문건수", ascending=False)
)

order_counts
```
