```python
import pandas as pd
df = pd.read_csv("02_Data.csv")
df
```

```python
# 1-1. 각 채널별 고객수 계산. 가장 많이 유입된 채널
# df.groupby("Channel").count()['Index'].idxmax()
most_channel = df['Channel'].value_counts().idxmax()
most_channel
```

```python
# 1-2. 가장 많이 유입된 채널에서 월렌탈비용의 평균
df[df['Channel'] == "영업방판"]['Amount_Month'].mean()
```

```python
# 1-3. 유입 경로가 대형마트A인 고객 중 연체 건수가 높은 상위 30명.
df2 = df[df['Channel'] == "대형마트A"]
df2 = df2.sort_values("Overdue_count", ascending = False).head(30)
df2.to_excel("result.xlsx", index = False)
```

```python
# 2-1. 고객들이 가장 많이 계약한 제품군 확인. 데이터 일부 출력.
idx = df['Product_Type'].value_counts().idxmax()
df[df['Product_Type'] == idx].head()
```

```python
# 2-2. 계약 건수가 가장 많은 상위 3개 제품군 추출
# 계약일자별 월렌탈비용의 합계.
items = df['Product_Type'].value_counts().head(3).index
df2 = df[df['Product_Type'].isin(items)]
df2.groupby("Datetime")['Amount_Month'].sum()
```

```python
# 2-3. 계약 건수가 가장 많은 상위 2개 제품군 간 월렌탈비용 평균 차이
items = df['Product_Type'].value_counts().head(2).index
A = df[df['Product_Type'].isin(items)].groupby("Product_Type")['Amount_Month'].mean()
abs(A.iloc[0] - A.iloc[1])
```

```python
# 3-1. 결제 은행이 신한카드인 고객 수 확인.
len(df[df['Bank'] == "신한카드"])
```

```python
# 3-2. 월렌탈비용이 15만원 이상이고, 판매 유형이 렌탈인 고객 수.
len(df[(df['Amount_Month'] >= 150000) & (df['Sales_Type'] == "렌탈")])
```

```python
# 4-1. 계약일자에서 계약 연도 파생 변수를 만들고
# 연도별, 제품군별 월렌탈비용의 합계
df['Datetime'] = pd.to_datetime(df['Datetime'])
df["year"] = df['Datetime'].dt.year
df.groupby(['year', 'Product_Type'])['Amount_Month'].sum()
```

```python
# 4-2. 제품군별 증감율 (2022년 대비 2023년)
df2 = df.pivot_table(index = "Product_Type", columns = "year",
              values = "Amount_Month", aggfunc="sum")
(df2[2023] - df2[2022]) / df2[2022]
```

```python
# 4-3. 판매 유형에서 렌탈과 멤버십의 월렌탈비용의 평균 차이
A = df.groupby("Sales_Type")['Amount_Month'].mean()
abs(A.iloc[0] - A.iloc[1])
```

```python
# 4-4. 주차별 계약 데이터 자동 정리 함수
# 계약일자를 기준으로 주차 데이터 생성
df["week"] = df['Datetime'].dt.isocalendar().week

def func(w):
    sample = df[df['week'] == w]
    result = sample.groupby("Product_Type")['Amount_Month'].agg(['mean', 'sum'])
    return result

func(5)
```
