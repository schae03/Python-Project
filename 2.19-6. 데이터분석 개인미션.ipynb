### 4. 전체 계약 건 분석

4-1. **계약 연도 파생 변수 생성 및 월렌탈비용 합계 계산**
- 계약일자(Datetime)에서 계약 연도(Year) 파생 변수를 생성하고, 연도별로 각 제품군(Product_Type)의 월렌탈비용(Amount_Month) 합계를 계산하시오.

```python
# 4-1
df["Year"] = pd.to_datetime(df["Datetime"]).dt.year 
df.groupby(["Year", "Product_Type"])["Amount_Month"].sum()
```

4-2. **제품군별 증감율 계산 (2022년 대비 2023년)**
- 2022년과 2023년 각 제품군(Product_Type)별 월렌탈비용(Amount_Month) 합계의 증감율을 계산하시오.
    - 증감율 계산식: 증감율=2022년 월렌탈비용 총합(2022년 월렌탈비용 총합)−(2023년 월렌탈비용 총합)

```python
# 4-2
# df.groupby(["Year", "Product_Type"])["Amount_Month"].sum()
A = (df[df["Year"] == 2022].groupby("Product_Type")["Amount_Month"].sum())
B = (df[df["Year"] == 2023].groupby("Product_Type")["Amount_Month"].sum())
print((A - B / A))
```

4-3. **판매 유형별 월렌탈비용 평균 차이 계산**
- 판매 유형(Sales_Type)에서 '렌탈' 계약과 '멤버십' 계약의 월렌탈비용(Amount_Month) 평균 차이를 계산하시오.

```python
# 4-3
print(df[df["Sales_Type"] == "렌탈"]["Amount_Month"].mean())       # 판매유형이 "렌탈"일때 월렌탈비용 평균 
print(df[df["Sales_Type"] == "멤버십"]["Amount_Month"].mean())     # 판매유형이 "멤버십"일때 월렌탈비용 평균 
print((df[df["Sales_Type"] == "렌탈"]["Amount_Month"].mean()) - (df[df["Sales_Type"] == "멤버십"]["Amount_Month"].mean()))

# 판매유형이 "렌탈"일때와 판매유형이 "멤버십"일때의 월렌탈비용의 평균 차이 : 34122.413318100276
```

4-4. **주차별 계약 데이터 자동 정리 함수 생성**
- 계약 데이터를 주기적으로 정리하는 업무 자동화 프로그램을 위해, 데이터에서 주차(Work Week)를 입력받아 해당 주차의 제품군(Product_Type)별 월렌탈비용(Amount_Month)의 합계와 평균을 엑셀 파일로 출력하는 함수를 생성하시오.

```python
# 4-4
def search_week(work_week):
    df['Datetime'] = pd.to_datetime(df['Datetime'])
    df['work_week'] = df['Datetime'].dt.to_period('W')
    df['week'] = df['work_week'].dt.week
    df_week = df[df['work_week'].dt.week == work_week]
    result = df_week.groupby('Product_Type')['Amount_Month'].agg(['sum', 'mean']).reset_index()
    print(result)
    return result.to_excel(f"{work_week}주차 월렌탈비용.xlsx")

A = input("주차를 입력하세요 : ")
search_week(int(A))
```
