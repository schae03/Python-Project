# 할로윈 투자 전략
할로윈 투자 전략은 매년 11월 초(할로윈 데이 다음 날인 11월 1일)에 주식을 매수하여 다음 해 4월 말까지 보유하고, 5월 초에 매도하여 10월 말까지는 현금이나 채권 등 안전자산을 보유하는 투자 전략을 말합니다. 즉, 11월부터 4월까지 6개월간은 주식 시장에 참여하고, 5월부터 10월까지 6개월간은 시장을 떠나 있는 것이 핵심입니다.

이 전략은 과거 수십 년간의 데이터를 분석했을 때, 11월-4월 기간의 주식 시장 수익률이 5월-10월 기간의 수익률보다 통계적으로 높게 나타났다는 경험적 관찰에 기반합니다.

```python
import FinanceDataReader as fdr
kospi = fdr.DataReader("KOSPI")
kospi
```

```python
kospi.loc["2001-11"]
```

```python
kospi.loc["2002-04"]
```

```python
cum_rate = 1

for year in range(2001, 2024):
    buy_mon = str(year) + '-11'
    sell_mon = str(year+1) + "-04"
    
    buy_price = kospi.loc[buy_mon].iloc[0]['Open']
    sell_price = kospi.loc[sell_mon].iloc[-1]['Close']

    rate = sell_price / buy_price
    cum_rate = cum_rate * rate

cum_rate # 약 5배 수익.
```

```python
# CAGR(연평균 성장률, Compound Annual Growth Rate)
# 연평균 수익률 약 7%
CAGR = (cum_rate ** (1/24)) - 1
CAGR
```

```python
# 단순보유누적수익률
simple_rate = kospi.iloc[-1]['Close'] / kospi.iloc[0]['Open']
simple_rate
```

```python
# 연평균 수익률 약 7%
simple_cum_rate = simple_rate ** (1/24) - 1
simple_cum_rate
```

```python
# 전략 백테스팅
import datetime
from dateutil.relativedelta import relativedelta

start = datetime.datetime(year = 2001, month = 11, day = 1)
start
```

```python
end = start + relativedelta(months = 5)
end
```

```python
start.strftime("%Y-%m")
```

```python
end.strftime("%Y-%m")
```

```python
def month6(df, start_year, end_year, month):

    cum_rate = 1
    
    for year in range(start_year, end_year+1):
        
        start = datetime.datetime(year=year, month=month, day = 1)
        end = start + relativedelta(months = 5)

        buy_mon = start.strftime("%Y-%m")
        sell_mon = end.strftime("%Y-%m")

        buy_price = df.loc[buy_mon].iloc[0]['Open']
        sell_price = df.loc[sell_mon].iloc[-1]['Close']

        rate = sell_price / buy_price
        cum_rate = cum_rate * rate

    return cum_rate
```

```python
# 누적 수익률 약 6.5배
month6(kospi, start_year = 2001, end_year = 2024, month = 11)
```

```python
data = {}
for month in range(1, 13):
    ret = month6(kospi, start_year = 2002, end_year = 2024, month = month)
    print(month, ret)
    data[month] = ret
```

```python
import matplotlib.pyplot as plt

plt.bar(data.keys(), data.values())
plt.xlabel("Month")
plt.ylabel("Return")
plt.grid(axis = 'y')
plt.xticks(range(1, 13))
plt.show()
```

4. KOSPI 지수 데이터를 사용하여 2002년부터 2023년까지 "역 할로윈 전략"의 누적 수익률을 계산하세요.

```python
cum_rate = 1
for year in range(2002, 2024):
    
    buy = str(year) + "-5"
    sell = str(year) + "-10"

    buy_price = kospi.loc[buy].iloc[0]['Open']
    sell_price = kospi.loc[sell].iloc[-1]['Close']
    
    rate = sell_price / buy_price
    cum_rate *= rate

cum_rate
# 0.73배 <= 손해본것
```
