```python
import pandas as pd

df1 = pd.read_excel("Data/07_ERP/기간별상품매출현황_202306.xlsx", header = 2).iloc[2:, 1:]
df2 = pd.read_excel("Data/07_ERP/기간별상품매출현황_202307.xlsx", header = 2).iloc[2:, 1:]
df3 = pd.read_excel("Data/07_ERP/기간별상품매출현황_202308.xlsx", header = 2).iloc[2:, 1:]
```

```python
df1['월'] = 6
df2['월'] = 7
df3['월'] = 8
```

```python
# 1. 기간(월) 컬럼 추가 및 " 합계" 컬럼 확인
import warnings
warnings.filterwarnings("ignore")

df = pd.concat([df1, df2, df3], ignore_index=True)
df = df.fillna(0)
df.to_excel("Data.xlsx", index = False)
del df[' 합계']
df
```

```python
# 2. 각 지점별 판매량 합계를 계산하고, 그 중 합계가 가장 높은 상위 2개 지점명을 추출
df2 = pd.melt(df, id_vars = ['상품코드', '상품명', "월"], var_name = "지점", value_name = "판매량")
top2 = df2.groupby("지점")['판매량'].sum().sort_values(ascending = False).head(2).index
top2
```

```python
# 3. 상위 2개 지점의 판매량 히스토그램 시각화
import seaborn as sns
import matplotlib.pyplot as plt
import koreanize_matplotlib

df3 = df2[df2['지점'].isin(top2)]

sns.histplot(data=df3, x = "판매량", hue = '지점')

plt.grid()
plt.show()
```

```python
# 4. 각 기간(월)별로 각 지점의 판매량 합계를 계산한 후, 최고 판매 지점을 확인
period_branch_sales = df2.groupby(['월', '지점'])['판매량'].sum().reset_index()
period_branch_sales
```

```python
max_branch_per_period = period_branch_sales.loc[period_branch_sales.groupby('월')['판매량'].idxmax()]
max_branch_per_period
```

```python
# 특정 월을 입력하면 해당 월의 각 지점별 판매량 합계를 엑셀 파일(result_{월}.xlsx)로 저장하는 함수
def save_branch_sales_by_period(month):
    data_month = df2[df2['월'] == month]
    branch_sales = data_month.groupby('지점')['판매량'].sum().reset_index()
    filename = f"result_{month}.xlsx"
    branch_sales.to_excel(filename, index=False)
    print(f"{month}월 데이터가 {filename} 파일로 저장되었습니다.")
```

```python
save_branch_sales_by_period(8)
```

```python
# 5. 정규성 검정 (상위 2개 지점의 판매량 데이터)

import scipy.stats as stats

for branch in top2:
    sales = df2[df2['지점'] == branch]['판매량']
    stat, p_value = stats.shapiro(sales)
    print(p_value)
```

```python
# 6. 가설 검정 (상위 2개 지점의 판매량 대표값 차이 비교)

# 상위 2개 지점의 판매량 데이터 추출
sales_branch1 = df2[df2['지점'] == top2[0]]['판매량']
sales_branch2 = df2[df2['지점'] == top2[1]]['판매량']
```

```python
# - 귀무가설(H0): 두 지점의 판매량 대표값(중앙값)에 차이가 없다.
# - 대립가설(H1): 두 지점의 판매량 대표값(중앙값)에 차이가 있다.
stats.mannwhitneyu(sales_branch1, sales_branch2)
```

```python
# 7. '상품명'을 기반으로 '마스크'와 '휴지'를 구분하고, 그 외 상품은 '기타'로 분류하여 파생 변수 '상품분류'를 생성
def classify_product(product_name):
    if '마스크' in product_name:
        return '마스크'
    elif '휴지' in product_name:
        return '휴지'
    else:
        return '기타'

df2['상품분류'] = df2['상품명'].apply(classify_product)
df2
```

```python
# '마스크'와 '휴지'의 판매량 데이터를 추출
mask_sales = df2[df2['상품분류'] == '마스크']['판매량']
tissue_sales = df2[df2['상품분류'] == '휴지']['판매량']
```

```python
# 1. 정규성 검정 (Shapiro-Wilk 검정)
stat_mask, p_mask = stats.shapiro(mask_sales)
stat_tissue, p_tissue = stats.shapiro(tissue_sales)

print(f"마스크 판매량 정규성 검정: W = {stat_mask:.3f}, p-value = {p_mask:.3f}")
print(f"휴지 판매량 정규성 검정: W = {stat_tissue:.3f}, p-value = {p_tissue:.3f}")
```

```python
stats.mannwhitneyu(mask_sales, tissue_sales)
```

```python
# 8. 제품 불량률 확률 계산 (불량률 10%, 30개 중 5개 이상 불량)
from scipy.stats import binom, poisson
1 - binom.cdf(4, 30, 0.1)
```

```python
# 9. 고객센터 문의 건수 확률 계산 (평균 3건, 하루 5건 이상 문의)
1 - poisson.cdf(4, 3)
```
