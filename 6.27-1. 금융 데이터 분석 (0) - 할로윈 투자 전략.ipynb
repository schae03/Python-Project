```python
# 1월에 사서 6월에 파는 것
# 2월에 사서 7월에 파는 것
# 11월에 사서 4월에 파는 것     -  이때가 더 좋지 않을까 추론

import FinanceDataReader as fdr
kospi = fdr.DataReader("KOSPI")
kospi
```

```python
# 할로윈 전략 적용 시 수익률 : 6.5배
cum_rate = 1

for year in range(2001, 2025):
    buy = str(year) + "-11"
    sell = str(year+1) + "-4"

    buy_price = kospi.loc[buy].iloc[0]["Open"]       # 11/1에 살려고 0번째인덱스에 접근
    sell_price = kospi.loc[sell].iloc[-1]["Close"]      # 4/30에 팔려고 -1(=마지막인덱스)에 접근

    rate = sell_price / buy_price
    cum_rate *= rate       # 누적해서 곱해줌

cum_rate      # 6.5배 수익률
```

```python
# 연평균 수익률 - 매년 8% 정도 수익
cum_rate ** (1/23) - 1
```

```python
# 단순보유누적수익률 : 5배
simple_rate = kospi.iloc[-1]["Close"] / kospi.iloc[0]["Open"]
simple_rate

# 위전략안취하고 존버하면 5배 수익률
```

```python
# 연평균 수익률 - 7% 정도 수익
simple_rate ** (1/23) - 1
```

```python
import datetime
from dateutil.relativedelta import relativedelta

start = datetime.datetime(year = 2001, month = 11, day = 1)
start
```

```python
end = start + relativedelta(months = 5)        # 5개월치만큼 더한다
end
```

```python
# 할로윈 전략 함수

def cal_months(df, start_year, end_year, month):
    cum_rate = 1
    for year in range(start_year, end_year + 1):
        start = datetime.datetime(year=year, month=month, day = 1)
        end = start + relativedelta(months = 5)

        buy = start.strftime("%Y-%m")
        sell = end.strftime("%Y-%m")
        
        buy_price = df.loc[buy].iloc[0]['Open']
        sell_price = df.loc[sell].iloc[-1]['Close']
        rate = sell_price / buy_price
        cum_rate *= rate
        
    return cum_rate
        
cal_months(kospi, 2001, 2024, 11)
```

```python
dic = {}
for month in range(1, 13):
    ret = cal_months(kospi, 2002, 2024, month)
    print(month, ret)
    dic[month] = ret

# 언제가 수익률이 가장 좋았는지 확인 가능
```

```python
import matplotlib.pyplot as plt

plt.bar(dic.keys(), dic.values())
plt.grid()
plt.xticks(range(1, 13))
plt.show()
```
