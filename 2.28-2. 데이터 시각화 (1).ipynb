# 데이터 시각화
데이터 시각화는 복잡한 데이터를 시각적 형태(예: 차트, 그래프, 지도 등)로 표현하여 데이터를 쉽게 이해하고 분석할 수 있도록 도와주는 기술입니다. 이를 통해 데이터의 패턴, 트렌드, 이상치 등을 직관적으로 파악할 수 있으며, 효과적인 의사결정 및 문제 해결에 기여합니다.

## 데이터 시각화의 주요 목적
* 정보 전달: 숫자나 텍스트로 표현된 데이터를 시각적인 요소로 변환해 누구나 쉽게 이해할 수 있도록 합니다.
* 통찰력 제공: 데이터 간의 관계, 패턴, 트렌드를 시각적으로 표현함으로써 숨겨진 의미나 인사이트를 도출할 수 있습니다.
* 의사결정 지원: 데이터를 시각화하면 비즈니스, 연구, 정책 결정 등 다양한 분야에서 보다 효과적인 결정을 내릴 수 있도록 도와줍니다.

## 데이터 시각화의 구성 요소
* 데이터 수집 및 처리: 시각화를 위해서는 우선 신뢰할 수 있는 데이터를 수집하고, 필요한 형태로 정제하는 과정이 필요합니다.
* 시각화 기법 선택: 데이터를 어떻게 표현할 것인지를 결정하는 단계로, 데이터의 특성과 분석 목적에 따라 막대그래프, 선그래프, 산점도, 파이 차트, 히트맵 등 다양한 시각화 기법이 활용됩니다.
* 도구 및 소프트웨어: 데이터 시각화를 수행하기 위해 Python의 matplotlib, seaborn, Plotly, R의 ggplot2, 또는 Tableau, Power BI와 같은 전문 소프트웨어를 사용할 수 있습니다.

## 데이터 시각화의 장점
* 직관적인 이해: 복잡한 데이터도 시각적으로 표현하면 한눈에 파악할 수 있어, 이해와 전달이 용이합니다.
* 패턴 인식: 데이터의 변화 추이와 관계를 시각적으로 확인함으로써 중요한 패턴이나 이상치를 빠르게 발견할 수 있습니다.
* 효율적인 커뮤니케이션: 시각화된 데이터는 팀 내외부의 커뮤니케이션에 효과적이며, 데이터 기반 스토리텔링을 가능하게 합니다.

데이터 시각화는 단순한 그래프나 차트 이상의 의미를 가지며, 데이터에서 중요한 인사이트를 도출하고 이를 기반으로 효과적인 의사결정을 지원하는 핵심 도구입니다. 데이터의 복잡성을 해소하고 보다 명확한 정보를 제공함으로써, 다양한 분야에서 그 활용도가 날로 증가하고 있습니다.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-5, 5, 0.1)          # range는 소수점 단위로 증가가 안되나, arange는 가능
y = np.sin(x)  # 사인 함수

plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(axis = "y")      # y값을 기준으로만 grid 그려줌 
plt.show()
```

* 축약형 옵션: 색상, 마커, 선 스타일을 한 번에 지정할 수 있는 축약형입니다.
    * 'ro-'는 빨간색(red, r) 원(marker, o)과 실선(line, -)을 의미합니다.

```python
#plt.plot(x, y, "ro--")   # 빨간색, 마커는 동그라미로
#plt.plot(x, y, linestyle = ":")   # 점선으로
#plt.plot(x, y, linewidth = 10)   # 선 굵기 조정 가능
plt.plot(x, y, marker = "*")    # 마커는 별표로 (마커종류 엄청 다양항)

plt.title("Line Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(axis = "y")      # y값을 기준으로만 grid 그려줌 
plt.show()
```

* linestyle: 기본값은 '-' (실선)이며, -- (대시선), -. (대시-점선), : (점선) 등이 있습니다.

```python
plt.plot(x, y, linestyle = ":")   # 점선으로

plt.title("Line Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(axis = "y")      # y값을 기준으로만 grid 그려줌 
plt.show()
```

* linewidth: 기본값은 보통 1.5 정도이며, 숫자로 지정합니다.

```python
plt.plot(x, y, linewidth = 10)   # 선 굵기 조정 가능

plt.title("Line Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(axis = "y")      # y값을 기준으로만 grid 그려줌 
plt.show()
```

* marker: 예) 'o' (원), 's' (사각형), '^' (삼각형) 등

```python
plt.plot(x, y, marker = "*")    # 마커는 별표로 (마커종류 엄청 다양항)

plt.title("Line Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(axis = "y")      # y값을 기준으로만 grid 그려줌 
plt.show()
```

* markersize: 마커의 크기를 지정합니다.

```python
plt.plot(x, y, marker = "s", markersize = 10)    # 마커는 별표로 (마커종류 엄청 다양항)

plt.title("Line Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(axis = "y")      # y값을 기준으로만 grid 그려줌 
plt.show()
```

* markerfacecolor: 마커 내부의 색상을 지정합니다.

```python
plt.plot(x, y, marker = "s", markersize = 10, 
         markerfacecolor = "red") 

plt.title("Line Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(axis = "y")      # y값을 기준으로만 grid 그려줌 
plt.show()
```

* markeredgecolor: 마커 테두리의 색상을 지정합니다.

```python
plt.plot(x, y, marker = "s", markersize = 10, 
         markerfacecolor = "red", markeredgecolor = "black") 

plt.title("Line Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(axis = "y")      # y값을 기준으로만 grid 그려줌 
plt.show()
```

* label: 범례(legend)를 사용할 때 선에 대한 설명을 추가합니다.

```python
plt.plot(x, y, marker = "s", markersize = 10, 
         markerfacecolor = "red", markeredgecolor = "black", label = "Test") 

plt.title("Line Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(axis = "y")
plt.legend()      # 범례 
plt.show()
```

* alpha: 선의 투명도를 지정합니다. 값은 0 (완전히 투명)에서 1 (불투명) 사이로 지정합니다.

```python
plt.plot(x, y, marker = "s", markersize = 10, 
         markerfacecolor = "red", markeredgecolor = "black", label = "Test", alpha = 0.5) 

plt.title("Line Plot")
plt.xlabel("X-label")
plt.ylabel("Y-label")
plt.grid(axis = "y")
plt.legend()      # 범례 
plt.show()
```

* x축과 y축 범위 지정하기: 데이터의 특정 부분만 확대해서 보고 싶을 때, 축의 범위를 직접 지정할 수 있습니다.

```python
plt.plot(x, y, marker = "s", markersize = 10, 
         markerfacecolor = "red", markeredgecolor = "black", label = "Test", alpha = 0.5) 

plt.title("Line Plot")
plt.xlabel("X-label")
plt.ylabel("Y-label")

plt.xlim(-2, 2)        # x축의 범위 -2~2까지만 표현하라 
plt.ylim(-2, 2)        # y축의 범위 -2~2까지만 표현하라 

plt.grid(axis = "y")
plt.legend()      # 범례 
plt.show()
```

* 여러 선 그리기: 한 그래프에 여러 선을 그리기 위해서는 plt.plot() 함수를 여러 번 호출하면 됩니다. 각 호출은 하나의 선을 추가하며, 각 선마다 다른 스타일이나 라벨을 부여할 수 있습니다.

```python
import koreanize_matplotlib

# 데이터 생성: 0부터 10 사이의 100개의 점
x = np.linspace(0, 10, 100)               # 10까지 범위에 포함.  간격 일정하게 100개 뽑아줌 
y1 = np.sin(x)   # 첫 번째 선: 사인 함수
y2 = np.cos(x)   # 두 번째 선: 코사인 함수

plt.plot(x, y1, label = "sin")
plt.plot(x, y2, label = "cos")
plt.legend()
plt.show()
```

* 여러 개의 자리 할당해서 그래프 그리기

```python
# 데이터 생성
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.sin(x) * np.cos(x)

fig, axes = plt.subplots(2, 2, figsize = (5, 5))      # 5x5로 가로2개세로2개 총 4개 도화지 할당 

axes[0,0].plot(x, y1, "r-")
axes[0,0].set_title("Data1")

axes[0,1].plot(x, y2, "b--")
axes[1,0].plot(x, y3, "g.-")
axes[1,1].plot(x, y4)

fig.tight_layout()      # 간격 적절하게 조정
plt.show()
```

2. 한 Figure에 sin(x)와 cos(x)를 다른 스타일로 그려 비교
    * x 데이터를 0부터 2π까지 100개 생성
    * y1 = sin(x), y2 = cos(x)를 계산
    * 서로 다른 색상 및 선 스타일을 사용하여 두 함수를 하나의 플롯에 그리기

```python
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label = "sin")
plt.plot(x, y2, label = "cos")
plt.legend()
plt.show()
```

3. 2행 2열의 서브플롯을 만들어 각각 다른 함수를 그려보기
    * 4개의 서브플롯에 각각 sin(x), cos(x), tan(x) (tan은 y범위를 제한할 것), 그리고 sin(x)*cos(x) 그래프를 그리기
    * 각 서브플롯에 제목과 축 라벨 추가
    * plt.tight_layout()을 이용해 서브플롯 간 간격 조정

```python
# 데이터 생성
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.sin(x) * np.cos(x)

fig, axes = plt.subplots(2, 2, figsize = (5, 5))      # 5x5로 가로2개세로2개 총 4개 도화지 할당 

axes[0,0].plot(x, y1, "r-")
axes[0,0].set_title("sin(x)")

axes[0,1].plot(x, y2, "b--")
axes[0,1].set_title("cos(x)")

axes[1,0].plot(x, y3, "g.-")
axes[1,0].set_xlabel("X")
axes[1,0].set_ylim(-10, 10)

axes[1,1].plot(x, y4)
axes[1,1].set_title("sin(x)*cos(x)")

plt.tight_layout()      # 간격 적절하게 조정
plt.show()
```

7. 매개변수를 이용한 3차원 선 그래프를 그려보기
    * t 변수를 0부터 10까지 100개 생성
    * x = sin(t), y = cos(t), z = t로 3D 선 그래프 그리기
    * ax.plot()를 사용하고, 축 라벨 및 제목 추가

```python
t = np.linspace(0, 10, 100)

x = np.sin(t)
y = np.cos(t)
z = t

fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot(111, projection = "3d")      # 3d창 만들어짐

ax.plot(x, y, z, color = "pink", marker = "h")
plt.title("매개변수를 이용한 3차원 선 그래프")
plt.xlabel("X")
plt.ylabel("Y")
plt.zlabel("Z")
plt.show()
```

9. 선 그래프의 색상, 선 스타일, 마커 등을 세밀하게 커스터마이징한 후, 이미지를 파일로 저장하기
    * x 데이터를 0부터 10까지, y 데이터를 임의의 함수(예: y = sin(x) + cos(x))로 생성
    * plt.plot()의 매개변수(color, linestyle, marker, linewidth 등)를 사용해 그래프 스타일 커스터마이징
    * plt.savefig('filename.png')를 이용해 그래프 이미지를 PNG 파일로 저장

```python
x = np.arange(0, 10, 0.1)          # range는 소수점 단위로 증가가 안되나, arange는 가능
y = np.sin(x) + np.cos(x)

plt.plot(x, y, color = "red", linestyle = ":", marker = "v", linewidth = "5")
plt.title("Line Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.savefig('lineplot.png')
plt.show()
```
