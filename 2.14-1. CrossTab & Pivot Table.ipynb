```python
# CrossTab: 내가 원하는 행과 열을 설정해서 빈도값

import pandas as pd
import seaborn as sns   # 데이터 시각화 라이브러리
tips = sns.load_dataset("tips")
tips.head()
```

```python
pd.crosstab(index = tips["sex"], columns = tips["day"],   # 행을 sex로, 열을 day로
            values = tips["tip"], aggfunc = "mean",      # 성별별로, 요일별로 tip의 총합 구해줌.
            margins = True, margins_name = "총계",       # 총몇개인지 총계까지 계산. 열이름도 바꿔줄수있음 
            normalize = "columns")                        # normalize = True로 하면 전체비율의 합이 1이 되게 비율로 계산해줌
```

```python
# Pivot Table: 내가 원하는 행과 열을 설정해서 집계값
import warnings
warnings.filterwarnings("ignore")

tips.pivot_table(index = "sex", columns = "day",          # 행을 sex로, 열을 day로
                values = ["tip", "total_bill"], aggfunc = "mean")      # 성별별로, 요일별로 tip의 평균값 구해줌.
                      # 여러값 볼수있음 (크로스탭은 안됨)
```

```python
tips.pivot_table(index = ["sex", "time"], columns = ["day"],          # 멀티인덱스,  멀티칼럼도 가능
                values = "tip", aggfunc = ["sum", "mean"],       # 집계값도 여러개 볼수있음
                 fill_value = 0, margins = True)
```

```python
# tip은 평균을 보고싶고 totalbill은 총합을 보고싶다

tips.pivot_table(index = ["sex", "time"], columns = ["day"],          # 멀티인덱스,  멀티칼럼도 가능
                values = ["tip", "total_bill"], 
                aggfunc = {"tip" : "mean", "total_bill" : "sum"},       # 집계값도 여러개 볼수있음
                fill_value = 0, margins = True)
```
