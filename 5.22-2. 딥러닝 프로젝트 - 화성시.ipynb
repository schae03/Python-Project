```python
# 날짜형식으로 변환
df['date'] = pd.to_datetime(df['ta_ymd'])

# 일자별 승인금액 합계 (전체 업종 기준)
daily_amt = df.groupby('date')['amt'].sum().reset_index()

# 시계열 그래프
import matplotlib.pyplot as plt
import koreanize_matplotlib

plt.figure(figsize=(14, 6))
plt.plot(daily_amt['date'], daily_amt['amt'], label='일별 승인금액 합계')
plt.title('일별 승인금액 추이')
plt.xlabel('날짜')
plt.ylabel('승인금액 합계')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 대기질 지표 및 소비 변수 간 상관관계 분석을 위한 전처리
# 일자별로 집계 (업종별 차이 제거를 위해 전체 합산)

df_daily = df.groupby('ta_ymd').agg({
    'amt': 'sum',
    'cnt': 'sum',
    'PM2.5': 'mean',
    'PM10': 'mean',
    'O3': 'mean',
    'NO2': 'mean',
    'CO': 'mean',
    'SO2': 'mean'
}).reset_index()

# 날짜형식 변환
df_daily['date'] = pd.to_datetime(df_daily['ta_ymd'])

# 상관계수 계산 (소비 변수 vs 대기질 변수)
correlation = df_daily[['amt', 'cnt', 'PM2.5', 'PM10', 'O3', 'NO2', 'CO', 'SO2']].corr()

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('소비 변수와 대기질 변수 간 상관관계')
plt.tight_layout()
plt.show()

correlation['amt'].sort_values(ascending=False)
```

```python
import statsmodels.api as sm

# 1️⃣ NO2 → amt 단순 선형 회귀 (전체 소비금액 기준)
X = df_daily[['NO2']]
y = df_daily['amt']

# 상수항 추가
X = sm.add_constant(X)

# 회귀모형 학습
model = sm.OLS(y, X).fit()
regression_summary = model.summary()

# 2️⃣ 대기질 악화 시 특정 업종 소비 변화 분석
# PM2.5 기준으로 대기질 악화일/정상일 구분
df['date'] = pd.to_datetime(df['ta_ymd'])
pm_threshold = df['PM2.5'].median()  # 중간값 기준

df['대기질_상태'] = df['PM2.5'].apply(lambda x: '나쁨' if x > pm_threshold else '보통')

# 업종별 대기질 상태에 따른 평균 승인금액 비교
pivot = df.groupby(['card_tpbuz_nm_2', '대기질_상태'])['amt'].mean().unstack()

# 변화율(%) 계산
pivot['증감률(%)'] = ((pivot['나쁨'] - pivot['보통']) / pivot['보통']) * 100

import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import numpy as np
import warnings
warnings.filterwarnings("ignore")

# 상위 변화 업종 시각화
pivot_sorted = pivot.sort_values('증감률(%)', ascending=False).dropna().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(data=pivot_sorted, x='증감률(%)', y=pivot_sorted.index, palette='coolwarm')
plt.axvline(x=0, color='gray', linestyle='--')
plt.title('대기질 악화 시 승인금액이 증가한 업종 TOP10')
plt.xlabel('대기질 나쁨 대비 승인금액 증감률 (%)')
plt.tight_layout()
plt.show()

regression_summary
```

```python
# 1️⃣ 다중 회귀 모델: NO2 + PM10 + PM2.5 → amt

X_multi = df_daily[['NO2', 'PM10', 'PM2.5']]
y_multi = df_daily['amt']
X_multi = sm.add_constant(X_multi)
multi_model = sm.OLS(y_multi, X_multi).fit()
multi_regression_summary = multi_model.summary()
multi_regression_summary
```

```python
# 2️⃣ 업종군별 시계열 클러스터링 (대기질에 민감한 업종군 탐색)
# 업종별 일자별 승인금액 피벗 테이블 생성
pivot_amt = df.pivot_table(index='date', columns='card_tpbuz_nm_2', values='amt', aggfunc='sum').fillna(0)

# 표준화
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

scaler = StandardScaler()
pivot_scaled = scaler.fit_transform(pivot_amt.T)

# 클러스터링 (k=4 기준)
kmeans = KMeans(n_clusters=2, random_state=42)
cluster_labels = kmeans.fit_predict(pivot_scaled)

# 업종별 클러스터 결과 저장
cluster_result = pd.DataFrame({
    '업종': pivot_amt.columns,
    '클러스터': cluster_labels
}).sort_values('클러스터')
cluster_result[cluster_result["클러스터"] == 0]
```

```python
# 업종별 시계열 Z-정규화 (StandardScaler를 업종별로 적용)
pivot_z = pivot_amt.copy()
pivot_z = pivot_z.apply(lambda x: (x - x.mean()) / x.std(), axis=0).fillna(0)

from sklearn.cluster import KMeans

pivot_scaled_z = pivot_z.T  # 업종 x 일자
kmeans_z = KMeans(n_clusters=3, random_state=42)
labels_z = kmeans_z.fit_predict(pivot_scaled_z)

# 결과 저장
cluster_result_z = pd.DataFrame({
    '업종': pivot_scaled_z.index,
    '클러스터': labels_z
}).sort_values('클러스터')
len(cluster_result_z[cluster_result_z["클러스터"] == 0]), len(cluster_result_z[cluster_result_z["클러스터"] == 1]), len(cluster_result_z[cluster_result_z["클러스터"] == 2])
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertias = []
for k in range(1, 10):
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(pivot_scaled_z)
    inertias.append(model.inertia_)

plt.plot(range(1, 10), inertias, marker='o')
plt.xlabel('클러스터 수 (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.grid(True)
plt.show()
```
