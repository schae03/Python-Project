# 대구 검색했을때 뉴스기사 10개의 제목

```python
import requests
url = requests.get("https://search.naver.com/search.naver?sm=tab_hty.top&where=news&ssc=tab.news.all&query=%EB%8C%80%EA%B5%AC&oquery=%EB%8C%80%EA%B5%AC&tqi=iJYK%2Flqpts0ssL6%2FI%2FZssssst1G-077103")
url
```

```python
from bs4 import BeautifulSoup   # html 소스코드를 파이썬으로 다룰 수 있게 변환

html = BeautifulSoup(url.text)      # url.text를 파이썬으로 다룰 수 있는 객체로 바꿔줘 
```

```python
news = html.find("ul", class_ = "list_news _infinite_list").find_all("li", class_ = "bx")
len(news)       # 10개의 뉴스기사를 찾았다는 뜻
```

```python
import pandas as pd

total = []
for i in news:
    a = i.find("a", class_ = "news_tit")
    title = a["title"].strip()
    link = a["href"]
    total.append([title, link])
    #print(a)

pd.DataFrame(total, columns = ["제목", "링크"])
```

# 연금복권 1등의 당첨조와 당첨번호

```python
import time

for n in range(1,21):
    url = requests.get(f"https://search.daum.net/search?w=tot&DA=JIM&rtmaxcoll=JIM&&q=%EC%97%B0%EA%B8%88%EB%B3%B5%EA%B6%8C%20720%2B%20{n}%ED%9A%8C%EC%B0%A8")
    html = BeautifulSoup(url.text)
    numbers = html.find("tr", class_ = "fst").text.split()[5:]
    total.append(numbers)
    print(f"{n}회 연금복권 데이터 저장 완료 : {numbers}")
    time.sleep(2)
```

```python
import pandas as pd
pd.DataFrame(total, columns = ["조", 1,2,3,4,5,6])
```
