```python
import pandas as pd
import lightgbm as lgb
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# 1. 날짜 처리
card_df['ta_ymd'] = pd.to_datetime(card_df['ta_ymd'])
weather_df['일시'] = pd.to_datetime(weather_df['일시'])

# 2. 카드데이터: 날짜+소비유형+카드 속성 기준으로 결제횟수 집계
grouped_card = (
    card_df.groupby(['ta_ymd', '소비유형', 'sex', 'age', 'day', 'weekend_or_holiday'])
    .agg({'cnt': 'sum', 'amt': 'sum'})
    .reset_index()
    .rename(columns={'cnt': '일별결제횟수'})
)

# 3. 전날 결제횟수 (소비유형 + 성별 + 연령대 기준)
grouped_card['cnt_lag1'] = grouped_card.groupby(['소비유형', 'sex', 'age'])['일별결제횟수'].shift(1)

# 4. 증감 타겟 생성
grouped_card['증감'] = (grouped_card['일별결제횟수'] > grouped_card['cnt_lag1']).astype(int)
grouped_card = grouped_card.dropna(subset=['cnt_lag1'])

# 5. 안산 날씨데이터 병합
weather_ansan = weather_df[weather_df['지점명'] == '안산'].copy()
merged = pd.merge(grouped_card, weather_ansan, left_on='ta_ymd', right_on='일시', how='left')
merged = merged.dropna()

# 소비유형을 숫자형 category로 변환
merged['소비유형_코드'] = LabelEncoder().fit_transform(merged['소비유형']).astype(int)

# 6. 입력 변수 설정 (카드 + 대기기상 변수 포함)
exclude_cols = ['ta_ymd', '일시', '소비유형', '증감', '지점', '지점명', "cnt_lag1", "일별결제횟수"]
feature_cols = [col for col in merged.columns if col not in exclude_cols]

X = merged[feature_cols].dropna()
y = merged.loc[X.index, '증감']

# category 지정 (소비유형_코드는 이미 int로 넣었기 때문에 생략 가능)
for col in ['sex', 'age', 'day', 'weekend_or_holiday']:
    if col in X.columns:
        X[col] = pd.Categorical(X[col])

# Train-test split + 학습
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)
model = lgb.LGBMClassifier(random_state=42)
model.fit(X_train, y_train)

import matplotlib.pyplot as plt
import pandas as pd

importance_df = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': model.feature_importances_
}).sort_values(by='Importance', ascending=True)

plt.figure(figsize=(10, 12))
plt.barh(importance_df['Feature'], importance_df['Importance'], color='mediumseagreen')
plt.xlabel("Importance")
plt.title("Feature Importance")
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import classification_report, accuracy_score

y_pred = model.predict(X_test)

print("✅ Accuracy:", accuracy_score(y_test, y_pred))
print("\n📊 상세 성능 보고서:")
print(classification_report(y_test, y_pred, target_names=["감소", "증가"]))
```

```python
import pandas as pd
import numpy as np
import lightgbm as lgb
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# 1. 이틀치 금액 기반으로 타겟 생성
grouped = (
    card_df.groupby(['ta_ymd', '소비유형', 'sex', 'age', 'day', 'weekend_or_holiday'])
    .agg({'amt': 'sum'})
    .reset_index()
    .rename(columns={'amt': '일별결제금액'})
)
grouped = grouped.sort_values(['소비유형', 'sex', 'age', 'ta_ymd'])

grouped['amt_lag1'] = grouped.groupby(['소비유형', 'sex', 'age'])['일별결제금액'].shift(1)
grouped['amt_lag2'] = grouped.groupby(['소비유형', 'sex', 'age'])['일별결제금액'].shift(2)

# 내일 금액이 오늘보다 증가했는가? (미래를 타겟으로 shift(-1))
grouped['target'] = (
    grouped.groupby(['소비유형', 'sex', 'age'])['일별결제금액'].shift(-1) > grouped['일별결제금액']
).astype(int)

grouped = grouped.dropna(subset=['amt_lag1', 'amt_lag2', 'target'])

# 2. 안산 날씨 데이터 병합
weather_df['일시'] = pd.to_datetime(weather_df['일시'])
weather_ansan = weather_df[weather_df['지점명'] == '안산']
grouped['ta_ymd'] = pd.to_datetime(grouped['ta_ymd'])
merged = pd.merge(grouped, weather_ansan, left_on='ta_ymd', right_on='일시', how='left')

# 3. 소비유형 인코딩
merged['소비유형_코드'] = LabelEncoder().fit_transform(merged['소비유형'])

# 4. 입력 변수 설정
exclude = ['ta_ymd', '일시', '소비유형', '지점', '지점명', '일별결제금액', 'target']
features = [col for col in merged.columns if col not in exclude]

X = merged[features].dropna()
y = merged.loc[X.index, 'target']

# 범주형 처리
for col in ['sex', 'age', 'day', 'weekend_or_holiday']:
    if col in X.columns:
        X[col] = pd.Categorical(X[col])

# 5. 모델 학습
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42, test_size=0.2)
model = lgb.LGBMClassifier(random_state=42)
model.fit(X_train, y_train)

# 6. Feature Importance 시각화
importances = model.feature_importances_
importance_df = pd.DataFrame({'Feature': X.columns, 'Importance': importances})
importance_df = importance_df.sort_values(by='Importance', ascending=True)

plt.figure(figsize=(10, 12))
plt.barh(importance_df['Feature'], importance_df['Importance'], color='mediumseagreen')
plt.xlabel("Importance")
plt.title("Feature Importance (어제+오늘 → 내일 금액 증감 예측)")
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import classification_report, accuracy_score

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred, target_names=["감소", "증가"]))
```

```python
import pandas as pd
import numpy as np
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

# 1. 집계
grouped = (
    card_df.groupby(['ta_ymd', '소비유형', 'sex', 'age', 'day', 'weekend_or_holiday'])
    .agg({'amt': 'sum'})
    .reset_index()
    .rename(columns={'amt': '일별결제금액'})
)
grouped = grouped.sort_values(['소비유형', 'sex', 'age', 'ta_ymd'])

# 2. lag + 타겟 생성
grouped['amt_lag1'] = grouped.groupby(['소비유형', 'sex', 'age'])['일별결제금액'].shift(1)
grouped['amt_lag2'] = grouped.groupby(['소비유형', 'sex', 'age'])['일별결제금액'].shift(2)
grouped['target_amt'] = grouped.groupby(['소비유형', 'sex', 'age'])['일별결제금액'].shift(-1)
grouped = grouped.dropna(subset=['amt_lag1', 'amt_lag2', 'target_amt'])

# 3. 날씨 병합
weather_df['일시'] = pd.to_datetime(weather_df['일시'])
weather_ansan = weather_df[weather_df['지점명'] == '안산']
grouped['ta_ymd'] = pd.to_datetime(grouped['ta_ymd'])
merged = pd.merge(grouped, weather_ansan, left_on='ta_ymd', right_on='일시', how='left')

# 4. 소비유형 코드
merged['소비유형_코드'] = LabelEncoder().fit_transform(merged['소비유형'])

# 5. 모델 입력 정의
exclude = ['ta_ymd', '일시', '소비유형', '지점', '지점명', '일별결제금액', 'target_amt']
features = [col for col in merged.columns if col not in exclude]
X = merged[features].dropna()
y = merged.loc[X.index, 'target_amt']

# 6. 범주형 처리
for col in ['sex', 'age', 'day', 'weekend_or_holiday']:
    if col in X.columns:
        X[col] = pd.Categorical(X[col])

# 7. 회귀 모델 학습
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
model = lgb.LGBMRegressor(random_state=42)
model.fit(X_train, y_train)

# 8. 성능 평가
y_pred = model.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"✅ RMSE (예측 vs 실제 결제금액): {rmse:,.0f}")

# 9. Feature Importance 시각화
importance_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': model.feature_importances_
}).sort_values(by='Importance', ascending=True)

plt.figure(figsize=(10, 12))
plt.barh(importance_df['Feature'], importance_df['Importance'], color='darkcyan')
plt.xlabel("Importance")
plt.title(f"Feature Importance (회귀모델 - 결제금액 예측, RMSE: {rmse:,.0f})")
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# 예측한 "내일 결제금액"이 실제보다 평균적으로 약 6,400만 원 정도 오차가 있다
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# === 1. 데이터 로딩 및 날짜 형식 변환 ===
card_df["ta_ymd"] = pd.to_datetime(card_df["ta_ymd"])
weather_df["일시"] = pd.to_datetime(weather_df["일시"])

# === 2. 대기/기상 정보 준비 ===
weather_use = weather_df[[
    "일시", "PM2.5_SI", "PM10_SI", "O3_SI", "NO2_SI", "CO_SI", "SO2_SI", "AQI", "평균기온(°C)", "일강수량(mm)"
]].copy()
weather_use.columns = ["ta_ymd", "PM2.5_SI", "PM10_SI", "O3_SI", "NO2_SI", "CO_SI", "SO2_SI", "AQI", "avg_temp", "precip"]

# === 3. 공휴일 여부 병합용 준비 ===
holiday_info = card_df[["ta_ymd", "weekend_or_holiday"]].drop_duplicates()

# === 4. 대상 소비유형 설정 ===
target_ages = [2, 3]
target_types = ["방어소비", "쾌락소비"]

for consume_type in target_types:
    print(f"\n--- 소비유형: {consume_type} ---")

    # 1. 데이터 필터링
    subset = card_df[
        (card_df["age"].isin(target_ages)) &
        (card_df["소비유형"] == consume_type)
    ].copy()

    if subset.empty:
        print("데이터 없음.")
        continue

    # 2. 일자별 amt 집계
    daily = subset.groupby("ta_ymd")["amt"].sum().reset_index()
    daily = daily.sort_values("ta_ymd")
    daily["prev_amt"] = daily["amt"].shift(1)
    daily["label"] = (daily["amt"] > daily["prev_amt"]).astype(int)
    daily = daily.dropna(subset=["prev_amt"])

    if daily["label"].nunique() < 2:
        print("라벨 다양성 부족. 스킵.")
        continue

    # 3. 대기질 + 기상 + 요일 + 공휴일 정보 merge
    df = pd.merge(daily, weather_use, on="ta_ymd", how="left")
    df["weekday"] = df["ta_ymd"].dt.weekday
    df = pd.merge(df, holiday_info, on="ta_ymd", how="left")

    # 4. 최종 X, y 구성
    X = df[[
        "PM2.5_SI", "PM10_SI", "O3_SI", "NO2_SI", "CO_SI", "SO2_SI", "AQI",
        "avg_temp", "precip", "weekday", "weekend_or_holiday"
    ]]
    y = df["label"]

    if len(X) < 100:
        print(f"데이터 너무 적음: {len(X)}개. 스킵.")
        continue

    # 5. 스케일링 및 분할
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    X_train, X_val, y_train, y_val = train_test_split(
        X_scaled, y, test_size=0.2, random_state=42, stratify=y
    )

    # 6. 모델 구성
    model = Sequential([
        Dense(32, activation="relu", input_shape=(X_train.shape[1],)),
        Dropout(0.2),
        Dense(16, activation="relu"),
        Dropout(0.2),
        Dense(1, activation="sigmoid")
    ])

    model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
    es = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=300,
        batch_size=32,
        callbacks=[es],
        verbose=0
    )

    # 7. 평가
    y_pred = (model.predict(X_val) > 0.5).astype(int)
    acc = accuracy_score(y_val, y_pred)
    print(f"Validation Accuracy: {acc:.4f}")
    
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.legend(['train', 'test'])
    plt.grid()
    plt.show()
```
