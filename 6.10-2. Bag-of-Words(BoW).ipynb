# Bag of Words (BoW) : 단어 가방
각 단어별로 몇번 나왔는지를 중요하게 생각 

```python
from sklearn.feature_extraction.text import CountVectorizer

corpus = [
    'He is a great great person.',
    'She is a great person.',
    'Python is a great programming language.',
    'Is Python a great great language?'
]

cv = CountVectorizer()
matrix = cv.fit_transform(corpus)
matrix

# 4개의 문장이 있었는데 단어가방을 만드니까 4행 8열짜리로 나왔다 
```

```python
cv.vocabulary_      # 단어들에 대해 컴퓨터가 라벨링
```

```python
import pandas as pd

pd.DataFrame(matrix.toarray(), columns = cv.get_feature_names_out())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from konlpy.tag import Okt

df = pd.read_excel("Data/text data.xlsx").dropna()

okt = Okt()

def tokenizer(x):
    return okt.morphs(x)

cv = CountVectorizer(tokenizer=tokenizer)

X = cv.fit_transform(df["document"])
X
```
