```python
import pandas as pd
pew = pd.read_csv("Data/pew.csv")
pew
```

```python
# 열에 있는 값들을 행으로 내릴 수 있음.
pd.melt(pew, id_vars = "religion",                      # id_vars : 내리지 않고 고정할 열 적어줌 
        var_name = "income", value_name = "count")             # var_name = 내려간 새로운 이름, value_name = 해당하는 값
```

```python
# 퀴즈3: weather.csv // 날씨 최저온도/최고온도
weather = pd.read_csv("Data/weather.csv")
weather
```

```python
weather2 = pd.melt(weather, id_vars = ["id", "year", "month", "element"],
       var_name = "Day", value_name = "temp")
weather2        # tmax와 tmin 때문에 month에 1이 2개, 2가 2개, ... => 행을 열로 올리려고 한다
```

```python
# 내가 원하는 행과 열을 설정해서 집계값을 보는 것  -  pivot_table
weather2.pivot_table(index = ["year","month","Day"],
                    columns = "element",
                    values = "temp", dropna = False)         # 기본으로 누락값 다 처리됨 (dropna = True)
```
