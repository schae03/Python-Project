```python
import pandas as pd
import numpy as np

df1 = pd.read_csv("Data/concat_1.csv")
df2 = pd.read_csv("Data/concat_2.csv")
df3 = pd.read_csv("Data/concat_3.csv")
```

```python
# concat함수

df = pd.concat([df1,df2,df3])     # default는 axis=0  ; 아랫방향으로 합침. 같은 열이름을 가진것끼리 합침
df    # 행번호 깔끔하지 않음.
```

```python
df = pd.concat([df1,df2,df3], ignore_index = True)      # 행번호 깔끔하게 하려면 ignore_index = True
df   
```

```python
df2.columns = ['A','C','E','F']
df2
```

```python
pd.concat([df1,df2,df3])         # 열이름 달라도 합쳐짐.  누락값으로 채워서라도
pd.concat([df1,df2,df3], join = "inner")      # 공통된 것만 합치고 싶으면 join = "inner"
```

```python
df = pd.concat([df1,df2,df3], axis = 1)        # axis = 1 있으므로 행이름을 기준으로 합침.  같은 열이름 여러개 존재할수있음
df = pd.concat([df1,df2,df3], axis = 1, ignore_index = True)        # 열이름 초기화 가능
df
```

```python
df2.index = [0,2,4,6]

pd.concat([df1,df2,df3], axis = 1)       # 행이름 달라도 누락값으로 채워서라도 합쳐짐
pd.concat([df1,df2,df3], axis = 1,  join = "inner")
```

```python
pd.concat([df1,df2,df3], keys = ['df11','df22','df33'])      # keys : 어떤 데이터프레임이 뭐였는지 왼쪽에 라벨링 붙여줌 
```
