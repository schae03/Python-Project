### 특정 시점(예: 2023년 1월)의 데이터를 기준으로 해당 업종이 디지털 채널을 주로 쓰는지 여부(디지털전환여부)를 예측

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import seaborn as sns

# ----------------------------------------
# 1. 데이터 불러오기
# ----------------------------------------
df = pd.read_csv("Data/최종_데이터.csv", encoding="cp949")
df["기준년월"] = df["기준년월"].astype(int)

# ----------------------------------------
# 2. 범주형 변수 인코딩
# ----------------------------------------
df["업종_중분류_코드"] = LabelEncoder().fit_transform(df["업종_중분류"])
df["사업장_시도_코드"] = LabelEncoder().fit_transform(df["사업장_시도"].fillna("미상"))

# ----------------------------------------
# 3. 좌수/개수/건수 범주형 수치화 (변수정의서 기반)
# ----------------------------------------
count_cols = [
    "요구불예금좌수", "거치식예금좌수", "적립식예금좌수", "수익증권좌수", "신탁좌수", "퇴직연금좌수",
    "여신_운전자금대출좌수", "여신_시설자금대출좌수", "신용카드개수"
]
txn_cols = [
    "외환_수출실적거래건수", "외환_수입실적거래건수", "창구거래건수", "인터넷뱅킹거래건수",
    "스마트뱅킹거래건수", "폰뱅킹거래건수", "ATM거래건수", "자동이체거래건수"
]
category_map = {
    '0개': 0, '0건': 0,
    '1개': 1, '1건': 1,
    '2개': 2, '2건': 2,
    '2개초과 5개이하': 3, '2건초과 5건이하': 3,
    '5개초과 10개이하': 4, '5건초과 10건이하': 4,
    '10개초과 20개이하': 5, '10건초과 20건이하': 5,
    '20개초과 30개이하': 6, '20건초과 30건이하': 6,
    '30개초과 40개이하': 7, '30건초과 40건이하': 7,
    '40개초과 50개이하': 8, '40건초과 50건이하': 8,
    '50개 초과': 9, '50건초과': 9
}
for col in count_cols + txn_cols:
    df[col] = df[col].map(category_map)

# ----------------------------------------
# 4. 디지털 채널 비율 및 전환 여부 라벨 생성
# ----------------------------------------
df["디지털_거래금액"] = df["스마트뱅킹거래금액"] + df["인터넷뱅킹거래금액"] + df["폰뱅킹거래금액"]
df["전통_거래금액"] = df["창구거래금액"] + df["ATM거래금액"]
df["전체_거래금액"] = df["디지털_거래금액"] + df["전통_거래금액"]
df = df[df["전체_거래금액"] > 0]

df["디지털_비율"] = df["디지털_거래금액"] / df["전체_거래금액"]
df["디지털전환여부"] = (df["디지털_비율"] >= 0.6).astype(int)

# ----------------------------------------
# 5. 시계열 기준 train/test 분리
# ----------------------------------------
train = df[df["기준년월"] < 202301].copy()
test = df[df["기준년월"] >= 202301].copy()

feature_cols = [
    "업종_중분류_코드", "사업장_시도_코드", "요구불예금잔액", "여신한도금액", "자동이체금액",
    "신용카드사용금액", "체크카드사용금액", "스마트뱅킹거래금액", "창구거래금액",
    "요구불예금좌수", "여신_운전자금대출좌수", "신용카드개수",
    "스마트뱅킹거래건수", "인터넷뱅킹거래건수", "ATM거래건수"
]

X_train = train[feature_cols]
y_train = train["디지털전환여부"]
X_test = test[feature_cols]
y_test = test["디지털전환여부"]

# ----------------------------------------
# 6. 모델 학습 및 예측
# ----------------------------------------
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# ----------------------------------------
# 7. 성능 평가 출력
# ----------------------------------------
print(classification_report(y_test, y_pred))

# ----------------------------------------
# 8. Feature Importance 시각화
# ----------------------------------------
importances = model.feature_importances_
importance_df = pd.DataFrame({
    "Feature": X_train.columns,
    "Importance": importances
}).sort_values(by="Importance", ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(data=importance_df, x="Importance", y="Feature", palette="viridis")
plt.title("Feature Importance (RandomForest)")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt

# 1. 데이터 불러오기 및 전처리
df = pd.read_csv("Data/최종_데이터.csv", encoding="cp949")
df["기준년월"] = pd.to_datetime(df["기준년월"].astype(str), format="%Y%m")
df.sort_values(by=["업종_중분류", "기준년월"], inplace=True)

# 2. 디지털 거래금액 생성
df["디지털_거래금액"] = df["스마트뱅킹거래금액"] + df["인터넷뱅킹거래금액"] + df["폰뱅킹거래금액"] + df["ATM거래금액"]

# 3. 특정 업종 선택 (예: 일반음식점업)
target_sector = '1차 금속 제조업'
df_sector = df[df["업종_중분류"] == target_sector].groupby("기준년월")["디지털_거래금액"].sum().reset_index()

# 4. 정규화 및 시계열 데이터 생성
scaler = MinMaxScaler()
scaled_series = scaler.fit_transform(df_sector[["디지털_거래금액"]])
seq_len = 6
X, y = [], []
for i in range(len(scaled_series) - seq_len):
    X.append(scaled_series[i:i+seq_len])
    y.append(scaled_series[i+seq_len])
X, y = np.array(X), np.array(y)

# 5. Train/Test 분할
split = int(len(X) * 0.8)
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# 6. LSTM 모델 정의
model = Sequential()
model.add(LSTM(64, input_shape=(seq_len, 1), activation="relu"))
model.add(Dense(1))
model.compile(optimizer="adam", loss="mse")

# 7. 학습
early_stop = EarlyStopping(monitor="val_loss", patience=10)
model.fit(X_train, y_train, epochs=100, validation_split=0.2, callbacks=[early_stop], verbose=0)

# 8. 예측 및 성능 확인
y_pred = model.predict(X_test)
y_test_inv = scaler.inverse_transform(y_test.reshape(-1, 1))
y_pred_inv = scaler.inverse_transform(y_pred)

mse = mean_squared_error(y_test_inv, y_pred_inv)
mae = mean_absolute_error(y_test_inv, y_pred_inv)
r2 = r2_score(y_test_inv, y_pred_inv)

print(f"MSE: {mse:.2f}, MAE: {mae:.2f}, R2: {r2:.3f}")

# 9. 결과 시각화
plt.plot(df_sector["기준년월"][-len(y_test):], y_test_inv, label="Actual")
plt.plot(df_sector["기준년월"][-len(y_test):], y_pred_inv, label="Predicted")
plt.title(f"{target_sector} 디지털 채널 이용금액 예측")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# 1. 데이터 로딩
df = pd.read_csv("Data/최종_데이터.csv", encoding="cp949")
df["기준년월"] = df["기준년월"].astype(int)

# 2. 디지털 거래금액 생성 및 필터링
df["디지털_거래금액"] = (
    df["스마트뱅킹거래금액"] + df["인터넷뱅킹거래금액"] + df["폰뱅킹거래금액"] + df["ATM거래금액"]
)
df = df[df["디지털_거래금액"] > 0]  # 0보다 큰 경우만 사용

# 3. 업종 필터링 (예: 도소매업)
sector = "종합 건설업"
df_sector = df[df["업종_중분류"] == sector].sort_values("기준년월")

# 4. 월별 디지털금액 집계
monthly = df_sector.groupby("기준년월")["디지털_거래금액"].sum().reset_index()

# 5. 로그 변환 및 스케일링
monthly["log_금액"] = np.log1p(monthly["디지털_거래금액"])
scaler = MinMaxScaler()
scaled = scaler.fit_transform(monthly[["log_금액"]])

# 6. 시계열 시퀀스 생성
def create_sequences(data, window_size):
    X, y = [], []
    for i in range(len(data) - window_size):
        X.append(data[i:i+window_size])
        y.append(data[i+window_size])
    return np.array(X), np.array(y)

window_size = 3
X, y = create_sequences(scaled, window_size)

# 7. 학습/테스트 분리
split_idx = int(len(X) * 0.8)
X_train, X_test = X[:split_idx], X[split_idx:]
y_train, y_test = y[:split_idx], y[split_idx:]

# 8. LSTM 모델 정의 및 학습
model = Sequential([
    LSTM(64, activation='tanh', return_sequences=True, input_shape=(window_size, 1)),
    Dropout(0.2),
    LSTM(32, activation='tanh'),
    Dense(1)
])
model.compile(optimizer='adam', loss='mse')
model.fit(X_train, y_train, epochs=100, batch_size=4, verbose=0)

# 9. 예측 및 역변환
y_pred = model.predict(X_test)
y_pred_inv = np.expm1(scaler.inverse_transform(y_pred))
y_test_inv = np.expm1(scaler.inverse_transform(y_test.reshape(-1, 1)))

# 10. 성능 평가
mse = mean_squared_error(y_test_inv, y_pred_inv)
mae = mean_absolute_error(y_test_inv, y_pred_inv)
r2 = r2_score(y_test_inv, y_pred_inv)

print(f"MSE: {mse:,.2f}, MAE: {mae:,.2f}, R2: {r2:.3f}")
```
