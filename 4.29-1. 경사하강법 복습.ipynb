# 경사하강법 복습. 이걸 활용해서 딥러닝 원리 알아볼것

```python
import numpy as np
import matplotlib.pyplot as plt
```

x = [2, 4, 6, 8]   
y = [20,60,55,80]

예측선 y^ = ax + b  <-  이 선만 알 수 있으면 공부시간에 따른 성적 예측할 수 있음   
오차가 얼마인지 알아야 학습가능.  e = Σ(y^ - y)^2 = Σ(ax + b - y)^2   

```python
X = np.array([2, 4, 6, 8])         # 공부시간
Y = np.array([20, 60, 55, 80])      # 정답지

a = 0
b = 0
lr = 0.005   # 학습률


for i in range(5000):
    y_hat = a * X + b
    error = y_hat - Y

    a_diff = sum(2 * error * X)     # a의 기울기값
    b_diff = sum(2 * error * 1)     # b의 기울기값
    
    a -= a_diff * lr
    b -= b_diff * lr

    print(f"{i+1}회 학습 / a기울기: {a_diff:.2f}, b기울기: {b_diff:.2f}")
    
```

```python
plt.scatter(X, Y)
plt.plot(X, a * X + b, "red")
plt.grid()
plt.show()
```

## x가 2개면 y^ = a1x1 + a2x2 + b

```python
X1 = np.array([2, 4, 6, 8])         # 공부시간
X2 = np.array([1, 2, 1, 3])        # 문제집 푼 개수
Y = np.array([20, 60, 55, 80])      # 정답지

a1 = 0
a2 = 0
b = 0
lr = 0.005   # 학습률


for i in range(5000):
    y_hat = a1 * X1 + a2 * X2 + b
    error = y_hat - Y

    a1_diff = sum(2 * error * X1)     # a1의 기울기값
    a2_diff = sum(2 * error * X2)     # a2의 기울기값
    b_diff = sum(2 * error * 1)     # b의 기울기값
    
    a1 -= a1_diff * lr
    a2 -= a2_diff * lr
    b -= b_diff * lr

    print(f"{i+1}회 학습 / a1기울기: {a1_diff:.2f}, a2기울기: {a2_diff:.2f}, b기울기: {b_diff:.2f}")
```

```python
plt.scatter(X1, Y)
plt.plot(X1, a1 * X1 + a2 * X2 + b, "red")
plt.grid()
plt.show()

# 직선보다는 좀 더 가까워짐
```

## sin값을 예측하는 2차 함수 선 그리기

```python
x = np.arange(-3, 3, 0.01)
y = np.sin(x)

a = 0
b = 0
c = 0
lr = 0.00005   # 학습률


for i in range(5000):
    y_hat = a * x**2 + b * x + c
    error = y_hat - y

    a_diff = sum(2 * error * x**2)     # a의 기울기값
    b_diff = sum(2 * error * x)     # a의 기울기값
    c_diff = sum(2 * error * 1)     # b의 기울기값
    
    a -= a_diff * lr
    b -= b_diff * lr
    c -= c_diff * lr

    print(f"{i+1}회 학습 / a기울기: {a_diff:.2f}, b기울기: {b_diff:.2f}, c기울기: {c_diff:.2f}")
```

```python
plt.scatter(x, y)
plt.plot(x, a * (x**2) + b * x + c, "red")
plt.grid()
plt.show()
```

## sin값을 예측하는 3차 함수 선 그리기

```python
x = np.arange(-3, 3, 0.01)
y = np.sin(x)

a = 0
b = 0
c = 0
d = 0
lr = 0.000005   # 학습률


for i in range(5000):
    y_hat = a * x**3 + b * x**2 + c * x + d
    error = y_hat - y

    a_diff = sum(2 * error * x**3)
    b_diff = sum(2 * error * x**2)
    c_diff = sum(2 * error * x)     
    d_diff = sum(2 * error * 1) 
    
    a -= a_diff * lr
    b -= b_diff * lr
    c -= c_diff * lr
    d -= d_diff * lr

    print(f"{i+1}회 학습 / a기울기: {a_diff:.2f}, b기울기: {b_diff:.2f}, c기울기: {c_diff:.2f}, d기울기: {d_diff:.2f}")
```

```python
plt.scatter(x, y)
plt.plot(x, a * x**3 + b * x**2 + c * x + d, "red")
plt.grid()
plt.show()
```
