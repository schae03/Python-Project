```python
import pandas as pd

visited = pd.read_csv("Data/visited.csv")
site = pd.read_csv("Data/site.csv")
```

```python
visited
```

```python
site
```

```python
# 데이터 병합  -  merge
A = visited.merge(site, left_on = "site",
              right_on = "name", how = "inner")               # default는 how = "inner" ; 공통적으로 있는 데이터만 merge해줌
del A['name']
A
```

```python
visited
```

```python
visited.merge(site, left_on = "site",
              right_on = "name", how = "outer")        # outer : 두 데이터 프레임의 정보 모두 있어야 함.

visited.merge(site, left_on = "site",
              right_on = "name", how = "left")          # left : 왼쪽에 있는 데이터 프레임(site)은 모두 있어야 함.

visited.merge(site, left_on = "site",
              right_on = "name", how = "right")         # right : 오른쪽에 있는 데이터 프레임(name)은 모두 있어야 함.
```

```python
left = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2']
}, index=['K0', 'K1', 'K2'])

right = pd.DataFrame({
    'C': ['C0', 'C1', 'C2'],
    'D': ['D0', 'D2', 'D3']} )

left
```

```python
right
```

```python
# 왼쪽의 기준은 index, 오른쪽의 기준은 행이름에 있는 특정한 값일 때
left.merge(right, left_index = True, right_on = "D", how = "inner")
```

```python
# 데이터 병합  -  pd.merge

# pd.merge(left,right)
left.merge(right, left_index = True,                # 행이름을 기준으로
          right_index = True, how = "inner")            # 기본값이 inner
```

```python
left.merge(right, left_index = True,
          right_index = True, how = "outer") 
```

```python
# 기준열의 키가 같을때  -  on = "key"

df1 = pd.DataFrame({
    'key': ['K0', 'K1'],
    'value': ['V0', 'V1']
})

df2 = pd.DataFrame({
    'key': ['K0', 'K1'],
    'value': ['V2', 'V3']
})
df1
```

```python
df2
```

```python
# df1.merge(df2, left_on = "key", right_on = "key")    # 기준열의 키가 같으면 굳이 이렇게 안해도됨
df1.merge(df2, on = "key")
df1.merge(df2, on = "key", suffixes = ("_left", "_right"))      # suffixes로 열 이름 지어줌 
```

```python
# 왼쪽에 있던건지, 오른쪽에 있던건지 확인  -  indicator = True

left = pd.DataFrame({
    'key': ['K0', 'K1', 'K2', 'K3'],
    'A': ['A0', 'A1', 'A2', 'A3']
})

right = pd.DataFrame({
    'key': ['K0', 'K1', 'K2', 'K4'],
    'B': ['B0', 'B1', 'B2', 'B4']
})

left
```

```python
right
```

```python
left.merge(right, on = "key", how = "outer",
           indicator = True)
```

```python
# 양쪽 모두 키 값이 유일한 경우 (일대일대응)  ;  validate = "one-to-one" 해서 에러가 안나면 일대일대응 
df_left = pd.DataFrame({
    'key': ['K0', 'K1', 'K2'],
    'A': [1, 2, 3]
})

df_right = pd.DataFrame({
    #'key': ['K0', 'K1', 'K2'],
    'key': ['K0', 'K0', 'K2'],
    'B': [4, 5, 6]
})
df_left
```

```python
df_right
```

```python
# 일대일대응   ex. 고객정보 / 고객상세정보
# df_left.merge(df_right, on = "key", validate = "one_to_one")

# 일대다대응   ex. 고객정보 / 고객 주문 내역
df_left.merge(df_right, on = "key", validate = "one_to_many")
```
