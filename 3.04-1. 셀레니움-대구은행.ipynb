# 100개 뉴스기사 수집 (제목, 그 링크 안에 있는 전체 본문내용) 

```python
# !pip install undetected-chromedriver
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

driver = uc.Chrome()
driver.get("https://search.naver.com/search.naver?ssc=tab.news.all&where=news&sm=tab_jum&query=%EB%8C%80%EA%B5%AC%EC%9D%80%ED%96%89")
```

```python
total = []

while True:
    news = driver.find_element(By.CSS_SELECTOR, "ul.list_news._infinite_list").find_elements(By.CSS_SELECTOR, "li.bx")
    for i in news:
        links = (i.find_elements(By.CSS_SELECTOR, "a.info"))
        if len(links) == 2:      # 원본기사링크랑 네이버기사링크 둘다 있는 경우
            title = (i.find_element(By.CSS_SELECTOR, 'a.news_tit').get_attribute("title"))
            link = links[-1].get_attribute("href")           # 두 링크 중 마지막 거 (= 네이버기사 링크)
            A = [title, link]       # 제목과 링크만 모은다
            if A not in total:
                total.append(A)
    body = driver.find_element(By.CSS_SELECTOR, "body")
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(2)
    print(f"{len(total)}개 뉴스기사 수집완료")
    if len(total) >= 100:
        break
```

```python
import pandas as pd
df = pd.DataFrame(total, columns = ["제목", "링크"])
df
```

```python
box = []
for i in df["링크"]:
    driver.implicitly_wait(10)
    driver.get(i)
    try:
        content = driver.find_element(By.CSS_SELECTOR, "div.newsct_article._article_body")
    except:
        content = driver.find_element(By.CSS_SELECTOR, "div._article_content")    # 스포츠 뉴스 기사의 경우
    box.append(content.text)
```

```python
df["내용"] = box
df.to_excel("대구은행.xlsx", index = False)
df
```
